# 1. 이진트리    

1. 각 노드 자식수 2 이하    
2. 포화이진트리 - > 각단계 꽉찬것.  완전이진트리 - > 앞부터 빈틈없이 채운것.
3. 높이 h인 포화이진트리의 노드수 2^h - 1  / N개의 노드 가진 완전 이진트리 높이 log(N+1)  
4. a[1]부터 사용,  a[i]의 부모 - > a[i//2],  a[i]의 자식 -> a[2i], a[2i+1] 
5. 편향될 경우 높이 높고, 메모리 낭비 심함 -> 균형트리 등장   






````python
class Node:
    def __init__(self,item,left=None, right=None):
        self.item = item 
        self.left = left   
        self.right = right   

        
class BinaryTree:
    def __init__(self): # 트리 생성자. 
        self.root = None


    def preorder(self,n): # 전위순회
        if n!=None:
            print(n.item)
            self.preorder(n.left)
            self.preorder(n.right)
            
    def inorder(self,n): # 중위순회 
        if n!=None:
            if n.left:
                self.inorder(n.left)
            print(n.item)
            if n.right:
                self.inorder(n.right)

            
    def postorder(self,n): # 후위순회 
        if n!=None:
            if n.left:
                self.postorder(n.left)
            if n.right:
                self.postorder(n.right)
            print(n.item)

        
    def levelorder(self, root): # 레벨순회 
        q = [] 
        q.append(root)
        
        while q:
            n = q.pop(0)
            print(n.item)
            
            if n.left:
                q.append(n.left)
            if n.right:
                q.append(n.right)
            

    
    def height(self,root): # 트리 높이 계산 
        if root == None:
            return 0  
        return max(self.height(root.left),self.height(root.right)) + 1 
        
        pass
    
    
t = BinaryTree()
n1 = Node(100)
n2 = Node(200)
n3 = Node(300)
n4 = Node(400)
n5 = Node(500)
n6 = Node(600)
n7 = Node(700)
n8 = Node(800)
n1.left = n2
n1.right = n3
n2.left = n4
n2.right = n5
n3.left = n6
n3.right = n7
n4.left = n8 
t.root = n1

print('\n전위순회:\t', end ="")
t.preorder(t.root)
print('\n중위순회:\t', end ="")
t.inorder(t.root)
print('\n후위순회:\t', end ="")
t.postorder(t.root)
print('\n레벨순회:\t', end ="")
t.levelorder(t.root)

print('트리높이=', t.height(t.root)) 


# 1. 수행시간. 각 연산들은 노드를 한번씩만 방문하므로 O(N)시간이 소요된다. 
#            
# 2.용도 : 힙트리 ->허프만코드 , AVL 트리,레드블랙트리, 스레드이진트리 
````

# 2. 이진힙    

1. 우선순위큐 를 구현하는 자료구조    
2. 우선순위큐 : 가장 높은우선순위 O(1) 접근,삭제 / 삽입되는 항목은 임의의 우선순위   
3. 힙:완전이진트리. 우선순위큐 위해만들어진것/ 반정렬상태 (부모노드키값 >자식키값 ) / 중복값허용    
4. 힙속성:  부모 우선순위 > 자식우선순위      
5. a[i]의 자식 a[2i],a[2i+1] / a[j]의 부모 a[j//2]   
6. 최소힙. 키값 작을수록 높은 우선순위  /    최대힙. 그반대   

````python
class BHeap:
    def __init__(self,a): # 이진힙 생성자 
        self.a = a   
        self.N = len(a) - 1 

    def create_heap(self) : # 초기 힙 만들기 heapq.heapify()와 동일 
        for i in range(self.N//2,0,-1):
            self.downheap(i)
        
    def insert(self, key_value): # 삽입연산
        self.N +=1
        self.a.append(key_value)
        self.upheap(self.N)
        
        
    def delete_min(self): # 최솟값 삭제
        if self.N == 0:
            print('힙 비어있음')
            return None
        
        minimum = self.a[1]
        self.a[1],self.a[-1] = self.a[-1], self.a[1]  
        del self.a[-1]
        self.N -=1  
        self.downheap(1)
        return minimum
    
    def downheap(self,i): # 힙 내려가며 힙속성 회복
        
        while 2*i <= self.N:
            k = 2*i   
            if k < self.N and self.a[k][0] > self.a[k+1][0]:
                k+=1
            
            if self.a[i][0] < self.a[k][0]:
                break   
                
            self.a[i],self.a[k] = self.a[k], self.a[i]
            i  = k   
            
    def upheap(self,j): #힙을 올라가며 힙속성 회복
        while j > 1 and self.a[j//2][0] > self.a[j][0]:  
            self.a[j],self.a[j//2] = self.a[j//2], self.a[j]
            j = j//2   

            
    def print_heap(self): # 힙 출력
        for i in range(1,self.N +1):
            print('[%2d]' % self.a[i][0], self.a[i][1], ']', end="")  
        print('\n힙의크기 = ', self.N)
````

````python
import heapq
heap = []
heapq.heappush(heap, 4)         # insert()
heapq.heappush(heap, 1)
heapq.heappush(heap, 7)
heapq.heappush(heap, 3)
print(heap)
     
heapq.heappop(heap)             # delete()
print(heap)
heapq.heappushpop(heap,7)    # item을 삽입한 후 delete_min() 수행  
print(heap)


print('@@@@@@@@@@@@@@@@@@@')

x  = [4, 1, 7, 3, 8, 5]
heapq.heapify(x)                # create_heap() 
print(x)

heapq.heapreplace(x,9)    # delete_min()먼저 수행 후, item
print(x)
````
