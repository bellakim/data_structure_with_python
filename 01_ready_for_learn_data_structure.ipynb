### 1.1 자료구조와 추상데이터타입
    

   >자료구조(Data Structure)는 일련의 동일한 데이터를 정돈하여 저장한 구성체를 의미한다.
   


데이터를 정돈하는 이유는 프로그램에 저장하는 데이터에 대해 탐색, 삽입, 삭제 등의 연산 효율적으로 수행하기 위해서이다.  
자료구조를 설계할 때에는 데이터와 데이터에 관련된 연산들을 함께 고려해서 설계헤애한다.


   >추상데이터타입(Abstract Data Type)은 데이터와 그 데이터에 대한 추상적인 연산들로써 구성된다. 
   
**추상데이터타입**은 자료구조를 프로그램으로 구현할때 데이터를 저장할 구조를 생성,저장데이터 처리하는 연산을 정의 하기위한 관계를 정형화 시킨 개념이다.  
(추상이란 연산을 구체적으로 어떻게 구현한다는 세부사항 포함 안한다는뜻.)  
```
     (추상 데이터타입)                  (자료구조) 
    데이터 + 관련연산    --[구현]-->   데이터 + 구체화된 관련연산 
```
**자료구조**는 추상데이터타입을 실제프로그램으로 구현한것을 의미한다.  
자료구조를 사용해 특정문제를 해결하기 위한 알고리즘 설계시에 데이터와 관련된 연산들의 관계를 개념적으로 정립하는 추상데이터타입에 기반하여 효율적으로 자료구조를 만들어야만 효율적인 알고리즘의 설계 가능하다.  

책에서다루는 자료구조들은 탐색, 삽입,삭제를 기본연산으로 가진다.

### 1.2 수행시간의 분석 

**수행시간** 으로 자료구조의의 효율성 측정할 수 있다.  
자료구조 에대한 연산 수행시간 측정방식은 알고리즘 성능 측정 방식과 같다.   
**시간복잡도** 는 알고리즘이 수행되는 시간을 나타내고  
**공간복잡도** 는 알고리즘이 수행되는 동안 사용되는 메모리 공간의 크기 를 나타낸다.  
(주로 시간 복잡도만으로 알고리즘 분석. 알고리즘들 대부분 비슷한 메모리공간 사용해서)  

알고리즘의 시간복잡도는 알고리즘이 실행되는 동안 사용된 기본적인 연산 횟수를 입력 크기의 함수로 나타낸다.  
> 기본연산(Elementary Operation)이란 데이터간 크기 비교, 데이터 읽기 및 갱신, 숫자 계산 등과 같은 단순한 연산을 의미한다.

알고리즘의 수행시간은 3가지 방법으로 분석한다.  
1. 최악경우 분석   -  (일반적으로 이것 사용)
2. 평균경우 분석   -  (입력으로 균등분포 가정해서 분석. 즉 입력 무작위가정) 
3. 최선경우 분석   -  (가장 빠른 수행시간, 최적알고리즘찾는데 활용)



### 1.3 수행시간의 점근표기법

**수행시간** 은 알고리즘이 수행하는 기본 연산 횟수를 입력 크기에 대한 함수로 표현한 것이다.  
기본 연산 횟수에 대한 함수는 다항식으로 표현되는데 이를 입력크기에 대한 함수로 표현하기 위해서 **점근표기법** 이 사용된다. 

>중요하지 않은 항과 상수 계수를 제거하면 이해를 방해하는 불필요한 부분이 없어져서 알고리즘의 실행 시간에서 중요한 부분인 성장률에 집중할 수 있습니다. 상수 계수와 중요하지 않은 항목을 제거한 것은 점근적 표기법(asymptotic notation)


            **3가지 점근 표기법**
1. O(Big-Oh)- 표기법 (최악의경우, 상한점근, 가장일반적으로 사용)
2. 오(Big-Omega)- 표기법 (빅오의 반대. 최선의경우, 하한점근)
3. 0 (Theta)- 표기법

> **1. O 표기의 정의**  
    모든 N>=N0에 대해서 f(N)<=cg(N)이 성립하는 양의 상수 c와 N0이 존재하면, f(N) = O(g(N)) 이다.

즉 f(N) = O(g(N))은 N0보다 큰 모든 N에 대해서  
f(N)이 양의 상수를 곱한 g(N)에 미치지 못한다는 뜻.  
g(N)은 f(N)의 상한.

ex) 한 알고리즘 수행시간 =  2N^2 + 3N + 5 라면 양의 **상수 c**(임의,최소일필요 x) 의 값을 최고 차항의 계수인 2보다 큰 4를 택하고 g(N) = N^2 으로 정하면, 3보다 큰 모든 N(임의, 최소일필요x)에 대하여 2N^2 + 3N + 5 < 4N^2 이 성립한다. 따라서 2N^2 + 3N + 5 = O(N^2) 이다.  
(2N^2 +3N+5 = O(N^3)도 성립하지만 g(N)선택시 보통 정의를 만족하는 가장 치수가 낮은 함수 선택한다.- 과장된 표현을 피하기 위해서)  

간단한 방법으로 g(N)찾는법은  
다항식에서 최고차항만을 취한 뒤 그 항의 계수를 제거하여 g(N)으로 정한다.


  
> **2. 오 표기의 정의**  
    모든 N>=N0에 대해서 f(N)>=cg(N)이 성립하는 양의 상수 c와 N0이 존재하면, f(N) = 오(g(N)) 이다.  
    
오 - 표기의 의미는 N0보다 큰 모든 N에 대해서 f(N)이 cg(N)보다 작지 않다는 뜻.  
즉 f(N) = 오(g(N))은 양의 상수를 곱한 g(N)이 f(N)에 미치지 못한다는 뜻.  
g(N)은 f(N)의 하한.  

ex) 알고리즘 수행시간 =  2N^2 + 3N + 5 일때, 양의 상수 c 값을 1로 택하고 g(N) = N^2 으로 정하면 1보다 큰 모든 N에 대해 2N^2+3N+5 > N^2이 성립한다. 따라서 2N^2 +3N+5 = 오(N^2)이다.  
(2N^2+3N+5=오(N)도 성립하지만 g(N) 선택시 가장 높은 차수의 함수를 선택한다.)



> **3. 0 표기의 정의**  
    모든 N>=N0에 대해서 c1g(N) <= f(N) <= c2g(N)이 성립하는 양의 상수 c1,c2,N0이 존재하면, f(N) = O(g(N)) 이다.  

0 표기는 수행시간의 O,오 표기가 동일한 경우에 사용한다.  
따라서 2N^2 + 3N +5 = O(N^2) , 오(N^2) 동시에 만족하므로 2N^2 + 3N +5 = 0(N^2) 이다. ( 0(N^2)는 N^2과 2N^2+3N+5가 유사한 증가율을 가지고 있다는 뜻.)  



    


### 1.4 파이썬 언어에 대한 기본적인 지식  

모르는것만 적어보겠음.  

파이썬의 리스트는 동적(Dynamic Array) 개념으로 구현되어 있다. 즉, 프로그램이 수행되는 도중에 리스트에 새 항목들이 추가되면 리스트의 크기를 확장하고 항목들의 삭제가 거듭되어 리스트의 항목 수가 줄어들면 리스트의 크기를 축소시킨다. 이러한 크기 조절은 파이썬에서 알아서 수행한다.  

리스트는 다음과 같이 메모리에 저장된다.  
```
[c] - > [ ] [ ] [ ] [ ] 
         l   l   l   l  
         40  10  70  60 
```
c[0]에는 40 이라는 정수 객체의 레퍼런스가 저장된다.  
c[1] = 40이 수행되면 다음과 같이 c[1]에 직접 정수가 저장되는 것이 아니라 40을 가리키는 주소 즉 레퍼런스가 저장된다.  

```
[c] - > [ ] [ ] [ ] [ ] 
         l  /    l   l  
         40     70  60 
```

### 1.5 순환  

   > 순환 : 함수의 실행 과정 중 스스로를 호출하는 것
   
순환으로 구현된 함수는 두부분으로 구성된다. 하나는 기본(Base)case이고 다른 하나는 순환 case이다. 기본 case는 스스로 더이상 호출 않는 부분. 순환 case는 스스로를 호출하는 부분. 함수 마지막부분에서 순환하는 것을 꼬리순환 이라고 한다.  

###### 예제 
```
남태평양에 있는 어느 나라에 11개의 섬이 그림 과 같이 다리로 연결되어 있다.이 나라의 관광청에서는 관광객들이 11개의 섬들을 모두 방문할 수 있는, 순서가 다른 3개의 관광코스를 만들었다. 각 코스의 관광은 섬 H에서 시작한다. 관광청에서는 각 관광코스의 방문 순서를 다음과 같은 규칙에 따라 만들었다. 
```

```
          H
       /    \
      F       S
     / \     /  \ 
    U   E    Z   K 
   /    /         \ 
  N     A          Y 
         \ 
           T

```

**A 코스** : 섬에 도착하면 항상 도착한 섬을 먼저 관광하고 그 다음엔 왼쪽 섬으로 관광을 진행한다.왼쪽 방향의 모든 섬들을 방문한 후에는 오른쪽 섬으로 관광을 진행한다. 

```
def A_course(n):           
    if n!= None:
        print(n.name, '->', end='') # 섬 N 방문
        A_course(n.left) 
        A_course(n.right) 
```
H->F-> U-> N ->E-> A-> T-> S-> Z-> K-> Y


**B 코스** : 도착한섬 관광 미루고 먼저 왼쪽섬으로 관광을 진행하고 왼쪽방향 모든 섬들 방문한 후에 돌아와서 섬 관광한다. 그 담음에 오른쪽 섬으로 관광을 진행한다. 
```
def B_course(n):           
    if n!= None:
        A_course(n.left)
        print(n.name, '->', end='') # 섬 N 방문
        A_course(n.right) 
```
N -> U -> F -> T -> A - > E -> H -> S -> K - > Y 


**C 코스** : 도착한섬 관광 미루고 먼저 왼쪽섬으로 관광을 진행하고 왼쪽방향 모든 섬들 방문한 후에 그 담음에 오른쪽 섬으로 관광을 진행한다. 돌아와서 섬 관광한다.
```
def C_course(n):           
    if n!= None:
        A_course(n.left)
        A_course(n.right)
        print(n.name, '->', end='') # 섬 N 방문
```  
N -> U -> T -> A -> E -> F -> Z -> Y -> K -> S -> H  






````python
# 프로그램 1-5 

class Node:
    def __init__(self, name, left=None, right=None): # 섬 생성자
        self.name = name 
        self.left = left
        self.right = right
        
def map(): # 지도만들기 
    n1 = Node('H')
    n2 = Node('F')
    n3 = Node('S')
    n4 = Node('U')
    n5 = Node('E')
    n6 = Node('Z')
    n7 = Node('K')
    n8 = Node('N')
    n9 = Node('A')
    n10 = Node('Y')
    n11 = Node('T')
    
    n1.left = n2
    n1.right = n3
    n2.left = n4
    n2.right = n5
    n3.left = n6
    n3.right = n7
    n4.left = n8
    n5.left = n9
    n7.right = n10
    n9.right = n11
    return n1      # 시작섬리턴

def A_course(n): # A- 코스
    if n != None:
        print(n.name ,'->', end = '')
        A_course(n.left)
        A_course(n.right)
        
def B_course(n):           
    if n!= None:
        B_course(n.left)
        print(n.name, '->', end='') # 섬 N 방문
        B_course(n.right) 
    
def C_course(n):           
    if n!= None:
        C_course(n.left)
        C_course(n.right)
        print(n.name, '->', end='') # 섬 N 방문
        
start = map()
print('A-코스:\t', end="")
A_course(start)
print('\nB-코스:\t', end=" ")
B_course(start)
print('\nC-코스:\t', end=" ")
C_course(start)
````

    오(Big-Omega)-표기: 점근적 하한
    0(Theta)-표기: 동일한 증가율  
6. 파이썬 언어는 객체지향 프로그래밍 언어로서 클래스를 선언하여 데이터를 객체에 저장하고 메소드를 선언하여 객체들에 대한 연산을 구현한다.  
7. 순환(Recursion)은 함수/메소드가 스스로를 호출하는 것이다.  
8. 순환으로 구현된 함수는 두 부분으로 구성된다.  
    기본(Base)case:스스로를 더 이상 호출하지 않는 부분
    순환 case: 스스로를 호출하는 부분  
9. 무한 호출을 방지하기 위해 선언한 변수 또는 수식의 값이 호출이 일어날 때마다 순환 case에서 감소되어 최종적으로 if 문의 조건에서 기본 case를 실행하도록 제어한다.  
10. 꼬리순환: 함수의 마지막 부분에서 순환하는 것. 꼬리 순환은 반복문으로 변환하는 것이 수행 속도와 메모리 사용 측면에서 효율적이다. 
11. 일반적으로 순환은 프로그램(알고리즘)의 가독성을 높일 수 있다는 장점을 갖지만, 시스템 스택을 사용하기 때문에 메모리 사용 측면에서는 비효율 적일 수도 있다.  
12. 순환으로 표현된 프로그램을 억지로 반복문으로 변환하는 경우 프로개래머가 시스템 스택의 수행을 처리해야 하므로 프로그램이 매우 복잡해지고, 반복문으로 변환된 프로그램의 수행 속도가 순환으로 구현된 프로그램보다 항상 빠르다는 보장도 없다.
