## 7.1 선택정렬  

```
리스트에서 아직 정렬 안된 부분의 원소들 중에서 최솟값을 선택하여 
정렬 안된 부분의 가장 왼쪽의 원소와 교환하는 정렬알고리즘. 
```  

<img src="images/" width="">   

주의할것은   
정렬된 부분의 키들은 오른쪽의 정렬 안된 부분의 키들보다 크지 않다.  
선택정렬은 항상 정렬 안된 부분에서 최솟값을 찾아 정렬 안된 부분의 가장 왼쪽원소와 교환해서.  
정렬안된부분이 마지막에 1개의 원소만 남게되면 정렬 마친다.  


````python
# 7-1 selection_sort.py  

# 내가한거  
# def selection_sort(a):
#     for index in range(len(a)):
#         min_value = a[index:][0]
#         print(a[index:])
        
#         for i, value in enumerate(a[index:]): # 나는 sub list탐색 을 그 sub_list의 0부터 시작.
#             if min_value > value:                         #나는 변수에 담은뒤에 비교 
#                 a[index],a[index+i] = value, min_value
#                 min_value = value
              
            
# 책에 있던 내용
# (여기선 나처럼 minimum을 변수로 안가지고 있고 인덱스만 가지고 있다가 값비교시 a[i] 이런식으로 바로 대입해서 비교했다. )
def selection_sort(a):
    for i in range(0,len(a)-1):
        minimum = i
        
        for j in range(i,len(a)): # sub list 탐색이 i부터 시작하는것 주의 
            if a[minimum] > a[j]:
                minimum = j
        a[i],a[minimum] = a[minimum], a[i]   # 최소 인덱스만 가지고 있다가~마지막에 for문 마지막에 교환  
        
        
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
selection_sort(a)
print('정렬 후:\t', end="")  
print(a)  

````

**수행시간**  
선택정렬은 루프가 1회 수행될 때마다 정렬 안된 부분에서 가장 작은 원소를 선택한다.  
따라서 최초 루프 수행될 때 N개의 원소 중 최솟값을 찾기 위해 N-1회의 원소 비교가 필요 
두번째  N-2회....
마찌맊으로 루푸 수행시 2개의 원소 비교하여 최솟값 찾는다.  
```

총비교 횟수는 

(N-1) + (N-2) + (N-3) +....2+1 = N(N-1)/2 = O(N^2)

```  
이미 정렬 되있거나 최악의 경우 역정렬상태에서도 항상 O(N^2) 수행시간 소요된다.  
선택정렬은 입력에 민감하지 않는 알고리즘.    
원소 교환하는 횟수가 N-1밖에 안되는 특징(정렬 알고리즘 중 가장 작은 최악 경우 원소 교환 횟수)  
효율성 많이 떨어져서 잘 안쓴다.  

## 7.2 삽입정렬   

```
삽입정렬은 리스트가 정렬된 부분과 정렬 안된 부분으로 나뉘며 정렬 안된부분의 가장 왼쪽 원소를 정렬된 부분에 삽입하는 방식의 정렬 알고리즘.
```  

<img src="images/" width="">     

정렬 안된 부분의 가장 왼쪽 원소(현재원소)를  
정렬된 부분의 원소들을 비교하여 알맞은 자리에 삽입한다.  
(현재 원소를 리스트 앞부분의 정렬 상태를 유지하며 삽입하는 과정을 반복한다.)  



````python
# 7-2 insertion_sort  

def insertion_sort(a):  
    for i in range(0,len(a)):   # 책에서는 1부터~~ 
        for j in range(i,0,-1):
            if a[j] < a[j-1]:
                a[j],a[j-1] = a[j-1],a[j]
                
        print(a)

a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
insertion_sort(a)
print('정렬 후:\t', end="")  
print(a) 
````

**수행시간**  
삽입정렬은 입력에 민감한 알고리즘.  
입력이 거의 정렬되어 있을 때 정렬이 매우 빠르고,(n-1번만 비교)  
입력이 역으로 정렬되어 있는 때에는 최악의 경우로 매우 느리다.(n(n-1)/2)  
(하지만 최악인 이경우에 데이터 교환수가 o(N^2)이므로 이땐 선택정렬이 좋은 성능을 보인다.)  

입력 데이터가 무작위인 경우, 삽입정렬의 평균의 경우 수행시간은?  
최종적으로 삽입되는 곳이 평균적으로 정렬된 부분의 중간이기 때문에  
1/2 x N(N-1)/2 ~= 1/4x N^2 = O(N^2)이다.  

**응용**  
이미 정렬된 파일 뒷부분에 소량의 신규 데이터 추가하여 정렬하는 경우.--->삽입정렬 우수한 성능 보인다.  
입력 크기 작은경우에도 우수.  



## 7.3 쉘정렬   

```
쉘정렬(shell sort)은 삽입정렬을 이용하여 작은 값을 가진 원솓들을 리스트의 앞부분으로 옮기며 
큰 값을 가진 원소들이 리스트의 뒷부분에 자리잡도록 만드는 과정을 반복하는 알고리즘이다.  
```  
삽입 정렬에 전처리과정(작은값-> 리스트 앞부분으로, 큰값-> 리스트 뒷부분으로) 을 추가한 것.  
기존 삽입정렬의 현재 원소를 앞부분에 삽입하기 위해 이웃하는 원소들끼리 비교하며 한 자리씩 이동하는 단접을 보완하기위해 고안.  
전처리 각 단계에서 **일정 간격으로 떨어진 원소들에 대해 삽입정렬을 수행**.

간격 h인 원소들끼리 정렬하는 것을 **h-정렬(h-sort)**이라고 한다.  

위 그림 h-정렬(4-정렬)이후  
완전 정렬 되지 않았지만 작은 숫자들(10,25,35)가 리스트 뒷부분, 큰 숫자들(95,90,80)이 뒷부분으로 이동했다.  

**쉘정렬은 h정렬의 h값(간격)을 줄여가며 정렬 수행**(마지막엔 간격 1로하여 정렬 완료)  

**@대표적 간격의 순서(h-sequence)@**  
```
Shell : N/2, N/4,.....1 (나누기 2->1될때까지)  
Hibbard : 2^k-1, 2^(k-1)-1 .....7,3,1
Knuth : (3^k-1)/2, ...,13,4,1  
Sedgewick : ... 109, 41, 19, 5, 1
Marcin Ciura : 1750, 701, 301, 132, 57, 23, 10 ,4 1 

```

````python
# 7-3 shell_sort.py  
# for 루프는 h 정렬 수행, h//=3으로 1/3 감소  

def shell_sort(a):
    h  = 4                       # 3x+1간격: 1, 4, 13,40, 121...중에서 4와 1만 사용.
    while(h >=1):
        for i in range(h, len(a)):  # h-정렬 수행.
            j = i
            while j >= h and a[j] < a[j-h]:
                a[j],a[j-h] = a[j-h],a[j]
                j -= h
                
        h //= 3  
            
                
        
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
shell_sort(a)
print('정렬 후:\t', end="")  
print(a) 
````

## 7.4 힙정렬  

(이거 4장에서 했던 내용인듯?)  

1. 키로 우선순위 정하는 최대힙 만든다.
2. 루트의 숫자를 힙의 마지막 노드의 숫자와 교환, 힙크기 1 감소  
3. 루트로 이동한 숫자로 인해 위배된 힙속성 downheap으로 복원  
4. 이과정 반복~ 나머지 원소들 정렬.

(최대힙 루트 a[0]말고 a[1]저장. 가장큰수 
-->이거 자식 인덱스 찾을때 2*i, 2*i+1이라 0곱하변 애매해져서 인듯.) 

````python
# 7-4 heap_sort.py  

def downheap(i, size): # 루트로 올라온 키에 대해 힙속성을 회복시킴   
    while 2*i <=size:
        k = 2*i
        # 좌 우중 승자 결정~승자하고 비교하겠다.
        if k < size and a[k] < a[k+1]:
            k+=1
        if a[i] >= a[k]: # 힙속성 만족하면 루프 나간다.
            break
        a[i], a[k] = a[k], a[i] # 자식원소가 더 크면 바꿔준다.
        i = k
        
def create_heap(a): # 정렬하기 전 최대 힙 만들기.
    hsize = len(a) -1
    for i in reversed(range(1,hsize//2+1)): # range(self.N//2,0, -1)
        downheap(i,hsize)
        
        
# 내 질문--> 그럴꺼면 아예 첨부터 하향식 힙 만들면 되는것 아닌가?  
# 내 답변--> 힙은 엄연히 말하면 헐렁하게 정렬된  
#           부모가 자식보다 크다는건 만족하지만 자식끼리 같은 촌수에있는것들끼리는 모른다. 
# 그렇기 때문에 여기서 해준것은 최대 값 = root에 있는 값이라는것은 어떤 자명하기 때문에  
# 이 값을 재일 뒤로 빼고 그 앞까지만 다운힙을 하는것을 
# 맨 끝부터~1까지 올라가면서 해나가자는것. 그렇게 하면   
# 일단 앞쪽은 같은 촌수 끼리는 정렬이 안되있을지 몰라도  
# 뒷쪽의 정렬되어져 가는곳은 정확히 정렬이 된 상태가 된다.  

def heap_sort(a):
    N = len(a) - 1
    for i in range(N):
        a[1],a[N] = a[N],a[1]  # 루트와 히ㅃ의 마찌맊 항몪 교환
        downheap(1,N-1)
        N -= 1                 # 힙 크기 1 감소  
    

    
a = [-1,54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
create_heap(a)
print('최대힙:\t', end="")
print(a)

heap_sort(a)
print('정렬 후:\t', end="")  
print(a) 
````
