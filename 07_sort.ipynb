## 7.1 선택정렬  

```
리스트에서 아직 정렬 안된 부분의 원소들 중에서 최솟값을 선택하여 
정렬 안된 부분의 가장 왼쪽의 원소와 교환하는 정렬알고리즘. 
```

주의할것은   
정렬된 부분의 키들은 오른쪽의 정렬 안된 부분의 키들보다 크지 않다.  
선택정렬은 항상 정렬 안된 부분에서 최솟값을 찾아 정렬 안된 부분의 가장 왼쪽원소와 교환해서.  
정렬안된부분이 마지막에 1개의 원소만 남게되면 정렬 마친다.  


`````python
# 7-1 selection_sort.py  

# 내가한거  
# def selection_sort(a):
#     for index in range(len(a)):
#         min_value = a[index:][0]
#         print(a[index:])
        
#         for i, value in enumerate(a[index:]): # 나는 sub list탐색 을 그 sub_list의 0부터 시작.
#             if min_value > value:                         #나는 변수에 담은뒤에 비교 
#                 a[index],a[index+i] = value, min_value
#                 min_value = value
              
            
# 책에 있던 내용
# (여기선 나처럼 minimum을 변수로 안가지고 있고 인덱스만 가지고 있다가 값비교시 a[i] 이런식으로 바로 대입해서 비교했다. )
def selection_sort(a):
    for i in range(0,len(a)-1):
        minimum = i
        
        for j in range(i,len(a)): # sub list 탐색이 i부터 시작하는것 주의 
            if a[minimum] > a[j]:
                minimum = j
        a[i],a[minimum] = a[minimum], a[i]   # 최소 인덱스만 가지고 있다가~마지막에 for문 마지막에 교환  
        
# 최소값 비교시 같다면 ? 같은것은 일단 미만 아니니 넘어감. -> 다음 회차에서 최소이니 어짜피 바로 왼쪽 으로 분류됨. 
        
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
selection_sort(a)
print('정렬 후:\t', end="")  
print(a)  

`````

**수행시간**  
선택정렬은 루프가 1회 수행될 때마다 정렬 안된 부분에서 가장 작은 원소를 선택한다.  
따라서 최초 루프 수행될 때 N개의 원소 중 최솟값을 찾기 위해 N-1회의 원소 비교가 필요 
두번째  N-2회....
마찌맊으로 루푸 수행시 2개의 원소 비교하여 최솟값 찾는다.  
```

총비교 횟수는 

(N-1) + (N-2) + (N-3) +....2+1 = N(N-1)/2 = O(N^2)

```  
이미 정렬 되있거나 최악의 경우 역정렬상태에서도 항상 O(N^2) 수행시간 소요된다.  
선택정렬은 입력에 민감하지 않는 알고리즘.    
원소 교환하는 횟수가 N-1밖에 안되는 특징(정렬 알고리즘 중 가장 작은 최악 경우 원소 교환 횟수)  
효율성 많이 떨어져서 잘 안쓴다.  

## 7.2 삽입정렬   

```
삽입정렬은 리스트가 정렬된 부분과 정렬 안된 부분으로 나뉘며 정렬 안된부분의 가장 왼쪽 원소를 정렬된 부분에 삽입하는 방식의 정렬 알고리즘.
```

정렬 안된 부분의 가장 왼쪽 원소(현재원소)를  
정렬된 부분의 원소들을 비교하여 알맞은 자리에 삽입한다.  
(현재 원소를 리스트 앞부분의 정렬 상태를 유지하며 삽입하는 과정을 반복한다.)  



`````python
# 7-2 insertion_sort  

def insertion_sort(a):  
    for i in range(0,len(a)):   # 책에서는 1부터~~ 
        for j in range(i,0,-1):     # 이게 -1이 아니라 0까지인 이유는 아래서 j, j-1 비교시
            if a[j] < a[j-1]:       # -1이라면 0, -1 즉 마지막것을 비교하게되서 꼬이게됨. / 0 이라면 1 ,0을 비교하니 괜찮다.            
                a[j],a[j-1] = a[j-1],a[j]            # j-1 조심 
            else:
                break                                # j-1 < j 이면 그앞으로는 볼것도없이 다 정렬된거라 break함 
                
        print(a)

a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
insertion_sort(a)
print('정렬 후:\t', end="")  
print(a) 
`````

## 7.3 쉘정렬   

```
쉘정렬(shell sort)은 삽입정렬을 이용하여 작은 값을 가진 원솓들을 리스트의 앞부분으로 옮기며 
큰 값을 가진 원소들이 리스트의 뒷부분에 자리잡도록 만드는 과정을 반복하는 알고리즘이다.  
```  
삽입 정렬에 전처리과정(작은값-> 리스트 앞부분으로, 큰값-> 리스트 뒷부분으로) 을 추가한 것.  
기존 삽입정렬의 현재 원소를 앞부분에 삽입하기 위해 이웃하는 원소들끼리 비교하며 한 자리씩 이동하는 단접을 보완하기위해 고안.  
전처리 각 단계에서 **일정 간격으로 떨어진 원소들에 대해 삽입정렬을 수행**.

간격 h인 원소들끼리 정렬하는 것을 **h-정렬(h-sort)**이라고 한다.  

위 그림 h-정렬(4-정렬)이후  
완전 정렬 되지 않았지만 작은 숫자들(10,25,35)가 리스트 뒷부분, 큰 숫자들(95,90,80)이 뒷부분으로 이동했다.  

**쉘정렬은 h정렬의 h값(간격)을 줄여가며 정렬 수행**  
(마지막엔 간격 1로하여 정렬 완료-> 이 앞까지기 전처리(각그룹 대소비교삽입정렬<자리바꾸기>). 여기부터 일반삽입정렬같다.)  

**@대표적 간격의 순서(h-sequence)@**  
```
Shell : N/2, N/4,.....1 (나누기 2->1될때까지)  
Hibbard : 2^k-1, 2^(k-1)-1 .....7,3,1
Knuth : (3^k-1)/2, ...,13,4,1  
Sedgewick : ... 109, 41, 19, 5, 1
Marcin Ciura : 1750, 701, 301, 132, 57, 23, 10 ,4 1 

```

`````python
# 7-3 shell_sort.py  
# for 루프는 h 정렬 수행, h//=3으로 1/3 감소  

def shell_sort(a):
    h  = 4                       # 3x+1간격: 1, 4, 13,40, 121...중에서 4와 1만 사용.
    while(h >=1):
        for i in range(h, len(a)):  # h-정렬 수행.
            j = i
            while j >= h and a[j] < a[j-h]:
                a[j],a[j-h] = a[j-h],a[j]
                j -= h
                
        h //= 3  
            
    # 대략 이런식임(사진참고) 
    # i(시작인덱스) 부터 비교 4간격 뒤로 시작~ 못하면 다음 i 
    # i=3 이전은 어짜피 그다음 간격 i=-1이니까 간격만큼 가서 시작. 
    # i=4)  4,0 비교 -> 위치교환    ->  0,-4 불가능. 정지 -> 그다음 i= 5     (뒷값 더 작으면 위치교환)
    # i=5)  5,1 비교 -> 위치교환    ->  1,-3 불가능, 정지 -> 그다음 i= 6 
    # ...................
    # i=9)  9,5 비교 -> 위치교환    ->  5,1 비교 -> 위치교환  -> 1, -3 불가능, 정지 -> 그다음 i= 10
    # i=10) 10,6 비교-> 위치교환    ->  6,2 비교 -> 위치교환  -> 2, -2 불가능, 정지 -> 그다음 i= 11  (while문 한번)
    # i=11)
    
    # 질문.자답 
    # 근데 이게 실질적으로 각 while문 하나가 하나의 의 h(4)에 대한 교환전부가 아니라 
    # h부터 시작 i 끝가지 대해 while문 각각 수행한게(위전체가) 하나의 h에 대한 교환 전부.(그림 7-5 한거)
    
        
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
shell_sort(a)
print('정렬 후:\t', end="")  
print(a) 
`````

`````python
# 7-4 heap_sort.py  

def downheap(i, size): # 루트로 올라온 키에 대해 힙속성을 회복시킴   
    while 2*i <=size:
        k = 2*i
        # 좌 우중 승자 결정~승자하고 비교하겠다.
        if k < size and a[k] < a[k+1]:
            k+=1
        if a[i] >= a[k]: # 힙속성 만족하면 루프 나간다.
            break
        a[i], a[k] = a[k], a[i] # 자식원소가 더 크면 바꿔준다.
        i = k
        
def create_heap(a): # 정렬하기 전 최대 힙 만들기.
    hsize = len(a) -1
    for i in reversed(range(1,hsize//2+1)): # range(self.N//2,0, -1)
        downheap(i,hsize)
        
    # N//2-> N//2-1-> N//2-2->... 1까지 올라가면서 
    # 각 노드에서 ~리프까지 경로상의 해당 노드 자식들이 부모보다 크다면 교환(다운힙)
    # ->이렇게하면 모든 노드가 자식보다 크다는 힙속성 만족 (기존의 무작위리스트 -> 최대힙완성)
    
        
        
# 내 질문--> 그럴꺼면 아예 첨부터 하향식 힙 만들면 되는것 아닌가?  
# 내 답변--> 힙은 엄연히 말하면 헐렁하게 정렬된  
#           부모가 자식보다 크다는건 만족하지만 자식끼리 같은 촌수에있는것들끼리는 모른다. 
# 그렇기 때문에 여기서 해준것은 최대 값 = root에 있는 값이라는것은 어떤 자명하기 때문에  
# 이 값을 재일 뒤로 빼고 그 앞까지만 다운힙을 하는것을 
# 맨 끝부터~1까지 올라가면서 해나가자는것. 그렇게 하면   
# 일단 앞쪽은 같은 촌수 끼리는 정렬이 안되있을지 몰라도  
# 뒷쪽의 정렬되어져 가는곳은 정확히 정렬이 된 상태가 된다.  

def heap_sort(a):
    N = len(a) - 1
    for i in range(N):
        a[1],a[N] = a[N],a[1]  # 루트와 히ㅃ의 마찌맊 항몪 교환
        downheap(1,N-1)   # 왜 -1이여야 하지 이미 위에서 len(a)-1해줬는데? ->아..마지막은 교환되서 최대값이 맨뒤에 N번째에 있는것. 
        N -= 1             # 힙 크기 1 감소  
                           #(이걸 해야 뒷부분은 추후 뒷부분
                           # <최대값부터 순차로 저장된부분>은 다운힙시 안건든다.)

    
a = [-1,54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
create_heap(a)
print('최대힙:\t', end="")
print(a)

heap_sort(a)
print('정렬 후:\t', end="")  
print(a) 
`````

## 7.5 합병정렬  

**합병정렬(Merge Sort)** 은 크기가 N인 입력을 1/2N 크기로 분할하고  
각각에 대해 재귀적으로 합병정렬을 수행한 후 2개의 각각 정렬된 부분을 정렬하는 알고리즘.  
(합병- 2개의 각각 정렬된 입력 합쳐서 하나의 정렬된 결과 얻는것.)

`````python
# 7-6 merge_srot.py  

def merge(a,b,low,mid,high):  
    
    i = low           # 앞부분 비교시작점 
    j = mid + 1       # 뒷부분 비교 시작점  
    
    for k in range(low,high+1):                #k = 저장 시작점. 
        #잔반처리 
        if i > mid:      # i 가 미드까지 올라왔다면  
            b[k] = a[j]     # 그이상 올라간다면 j 의시작점.
            j+=1            # 걍 j 증가시키고 저장해주면됨 
                
        elif j > high:
            b[k] = a[i]
            i+=1
        
        
        elif a[i] <= a[j]:
            b[k] = a[i]
            i+=1

        elif a[i] > a[j]:
            b[k] = a[j]
            j+=1
            
    # b를 low부터 high 까지 a에게 최신화 시켜주자.  
    for k in range(low,high+1):
        a[k] = b[k]
    

def merge_sort(a,b,low,high):  
    
    if low >= high: # 원소 1개 이하 정렬시키려하면 걍 리턴함.
        print(low,high)
        return      # 우리는 a에 정렬이 바로 반영되므로(b거치긴하지만) 
                    # 정렬되있다면(원소한개라 정렬필요없다면) 그냥 리턴만 해도됨. 
        
    mid = (low+high)//2
    merge_sort(a,b,low,mid)      # 앞부분 정렬 시킴 
    merge_sort(a,b,mid+1,high)   # 뒷부분 정렬 시킴 
    merge(a,b,low,mid,high)      # 앞 뒷부분 합병
                                 # (두 시작점 i,j부터 각각 -> mid, high까지 타고가면서                      
                                 # 더작은 것을 b에저장후 마지막에 a에 반영함.                      
            
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]
b = [None]*len(a)
print('정렬 전:\t', end="")
print(a)
merge_sort(a,b,0,len(a)-1)
print('정렬 후:\t', end="")  
print(a) 
`````

## 7.6 퀵 정렬  
퀵정렬은 맨 윗쪽 원소(피벗,pivot)를 기준으로  
피벗보다 작은 원소들과 큰 원소들을 각각 피벗의 좌우로 분할한 후  
피벗보다 작은부분, 큰부분을 각각 재귀적으로 정렬하는 알고리즘

`````python
# 모두의 알고리즘에서 했던것 상기해서 먼저 만들어보자.  

# 그룹을 둘러 나눠 재귀 호출 하는 방식은 병합정렬과 같지만
# 그 그룹을 나눌 때 미리 기준과 비교해서 나눈다는 점이 다르다. 
# 즉. 먼저 기준과 비교를 해서 그룹을 나눈 다음 각각 제귀 호출하여 합치는 방식.



# 일단 부분 리스트의 맨 앞을 항상 피벗으로 가져가는것 같다.  
def qsort_modu(a,pivot,high):
    
    if len(a) <=1:
        return a
    
    group1 = list()
    group2 = list()  
    
    for index in range(pivot+1,high+1):
        if a[index] < a[pivot]:
            group1.append(a[index])
        else:
            group2.append(a[index])
    
    # 두 그룹에 대해서도 qsort 호출  
    group1 = qsort_modu(group1,0,len(group1)-1)
    group2 = qsort_modu(group2,0,len(group2)-1)

    
    # 피벗 기준으로 작은,큰 집단 a에 다시 합침
    middle = [a[pivot]]
    result = list()
    if group1:
        result = group1
    result  = result + middle
    
    if group2:
        result =result + group2
        
    
    return result 
    
    

a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]
print('정렬 전:\t', end="")
print(a)
a = qsort_modu(a,0,len(a)-1)
print('정렬 후:\t', end="")  
print(a) 
`````

`````python
# 원래 모두의 알고리즘 있던 내용 복붙  

def quick_sort_sub(a,start,end):
    # 종료조건: 정렬 대상이 한 개 이하이면 정렬할 필요가 없음
    if end - start <=0:
        return
    # 기준 값을 정하고 기준 값에 맞춰 리스트 안에서 각 자료의 위치를 맞춤
    # [기준 값보다 작은 값들, 기준값, 기준 값보다 큰 값들]
    pivot = a[end]
    i = start
    
    for j in range(start,end):
        if a[j] <= pivot:
            a[i],a[j] = a[j],a[i]               # i 가 j 보다 앞설수는 없어서 이런식으로 교환 가능한듯. 
            i+=1                                # j는 항상 한번도 교환 안된 인덱스. i부터 차근차근 넣어가면 되는것.
    a[i], a[end] = a[end],a[i]                  # 이미 i에 알맞은(피벗보다 작은값) 들어가 있을 수는 없다. 
    #재귀 호출 부분                                # 그런 경우가 있다면 i위치가 j위치가 같을때 뿐이고 의미없는 a[i] = a[j] 할뿐
    quick_sort_sub(a,start, i-1) #기준 값보다 작은 그룹을 재귀 호출로 다시 정렬 
    quick_sort_sub(a,i+1, end) # 기준 값보다 큰 그룹을 재귀 호출로 다시 정렬
    
# 리스트 전체(0~len(a)-1)를 대상으로 재귀 호출 함수
def quick_sort(a):
    quick_sort_sub(a,0,len(a)-1)
        
d = [6,8,3,9,10,1,2,4,7,5]
quick_sort(d)
print(d)


# 이경우 생각해보자. [ 5,2,3,4,    7,8,9,1,    6] 
# 피벗이 6인경우   
# 이경우 앞의 5,2,3,4걍 의미없는 교환하고 
# 789지나가고 
# 1만 7과 바꾼뒤 종료한다. ------->위에 코드 문제 없어보임.  
`````

`````python
#7.7 quick_sort.py ( 자료구조 책 본문)

# 기존의 모두의 알고리즘에서는 i-> 새롭게 저장되는 곳의 인덱스 (작은값만)           <i값 저장될때마다 전진>
#                         j-> 전체 index돌면서 피벗값과 비교해 나가는 인덱스   <끝가지 계속 한칸씩 전진>
#                         

# 여기서는 비슷한데(i에 작은값 새로이저장, j에 큰값 이동이라는 방식은 같지만) 
# i,j 가 둘다 큰 while문 안에서 중앙으로 향한다. 
# 각각의 i ,j 에 대한 while에서는 
# 자신의 그룹속성(피벗보다 각각 작고,크다 는) 에 위반되면  
# i,j인덱스 증가, 감소를 멈추고   
# 해당 인덱의 원소를 바꿔야 앞의 그룹과 바꿔야 되는 것이기에 바꾼후 
# 거기부터 다시 그다음 턴의 while 루프 코드에 진입해서 
# 다시 위반되는 값을 찾는다. 
# 그러다가 결국
# i 증가하다 j의 영역혹은 중립(아직 구분안된) 침범, j는 i혹은 중립영역 침범하게되고 
# j<=i 이 조건에 걸려 큰 while문 종료된다.  
# 이경우 
# 이시점에서 j는 i 의 영역 침범, i는 j영역 침범한 상태 
# (왜냐면 각각의 while조건에 의해 어떻게든 피벗보다 작으,크면 전진해서 위반되는 상태에서 멈추게 되니까.)
# j가 그시점에 있는 인덱스가 피벗이 있을 위치이므로 
# 맨앞의 피벗과 바꾼다. (근데 만약 초기에 피벗을 첫값이 아니라 끝값으로 잡았다면 i의 위치와 바꿔야 된다.)  

def qsort(a,low,high):
    if low < high:
        pivot = partition(a,low,high) # 피벗기준 분할 
        qsort(a,low,pivot-1)  
        qsort(a,pivot+1,high)   # 앞, 뒤부분 재귀호출 
    print(a)
        
def partition(a,pivot,high):
    i = pivot + 1 
    j = high
    while True:           # i<high, j>pivot 조건 궂이 쓸 필요 있나? 
        while i<high and a[i]<a[pivot]:  # a[i]가 피벗보다 작으면 i를 1 증가.  
            i+=1
        while j>pivot and a[j]>a[pivot]: # a[j]가 피벗보다 크면 j를 1 감소 
            j-=1
        if j<=i:                         # 루프중단
            break
        
        a[i],a[j] = a[j],a[i]            # a[i]와 a[j] 교환
        i += 1                            
        j -= 1                           # 교환하게 되면 해당 인덱스,i,j에서 둘다 그룹속성 만족하니까.
                                         # 각자 중앙으로 전진 
        
    a[j],a[pivot] = a[pivot],a[j]        # a[j]와 피벗 교환 
    return j                             # 피벗 인덱스 

a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]
print('정렬 전:\t', end="")
print(a)
qsort(a,0,len(a)-1)
print('정렬 후:\t', end="")  
print(a) 
`````

(ex) 10진수 의 기 : 0~9  
**LSD**(Least Significant Digit) : 자릿수 비교를 최하위숫자 ~취상위 숫자 방향으로 정렬   
**MSD**(Most Significant Digit) : 최상위~ 최하위 숫자 방향으로 정렬

## 7.8 외부정렬   

실세계의 대용량 데이터는 보조기억 장치(하드)에 저장 -> 내부정렬 만으로는 어렵다.  

**외부정렬** : 보조기억장치에 있는 대용량의 데이터를 정렬하는것.  
             합병절렬 사용   
             수행시간: 입력 전체를 **처리하는 횟수**( 원소의 비교 횟수 x)  
                     (보조 기억장치 접근시간이 메인메모리 접근 시간보다 매우 느려서)
                     
**패스(Pass)** : 입력(또는 데이터) 전체 처리하는 단위.
                (ex 컴파일러 = Two pass 컴파일러. 첫패스에서 소스프로그램 ->어셈블리어 변환 
                                              두번째패스에서 어셈블리어 -> 기계어 변환  )  
                                              

**ex)**  
메인메모리 데이터 정렬 용량 : 1GB  
디스크 에 저장된 입력  : 64GB  

1. 디스크로부터 메인 메모리에서 수용할 만큼의 입력 (1GB) 읽어 들인다.   
2. 퀵정렬과 같은 내부 정렬 알고리즘 사용해서 정렬   
3. 그 결과를 디스크에 일단 다시 저장.  
4. 이 과정 반복  
5. 기존의 64GB입력 -> 64개의 정렬된 블록으로 분할, 디스크에 저장된 상태   
   (**Run**: 정렬된 블록(데이터))  
6. 정렬된 블록들을 반복적 합병 -> 하나의 정렬된 블록(64GB)만든다.  
7. (이하 합병 상세과정) 입력 버퍼가 k 개 있으므로 KGB 블록이 총 64/k 개 만들어짐  
8. KGB블록을 k개씩 짝지어 합병시키면 k^2 GB 크끼의 블록이 64/k^2개 만들어진다.  
9. 이과정 반복하여 합병 진행   
10. 할때마다 블록크기 k배, 블록수 1/k로 줄어들어  64GB크기 블록 하나만 남는다.  

풀어쓰면 블록들을 부분적으로 메인 메모리의 입력 버퍼(Buffer)에 읽어들여서   
합병을 수행하여 부분적으로 출력 버퍼를 통해 디스크에 쓰는 과정 반복  

<img src="images/20181218_213812_868.jpg" width="800">  
(내부정렬로 만든 64개의 정렬된 블록)  
<img src="images/20181218_213819_804.jpg" width="800">    
(k개의 버퍼 이용한 합병: 1GB 블록들을 부분적으로 k개의 입력 버퍼로 읽어들여 kGB크기의 블록 만드는 과정)   


(예제)  
위 그림에서 k =  2이면?  
첫 pass -> 64/2= 32개의  2GB블록   
두번pass -> 32/2= 64/2^2 = 16개의 4(=2^2)GB 블록   
세번pass -> 16/2= 64/2^3 = 8개의 8GB블록  
....
여섯pass -> 64/2^6 = 1개의 64GB 블록  


**수행시간**  
```
입력크기 N
메인메모리 데이터 정렬 가능 용량 M
입력버퍼 k

첫 pass -> N/M개 블록 생성

k개의 블록을 하나의 블록으로 합병하는 방식~  

두 pass - > (N/M)/k
세 pass - > (N/M)/k^2
....

(N/M)/k^n = 1되는 n = log (N/M)
                         k  
                         
첫번째 pass더하면  log (N/M) +1 회의 Pass가 정렬을 위해 필요  
                   k
```

## 4번동영상  merging runs optimally with galloping      
[영상1](https://youtu.be/o8y9uYygLcw)   


정렬된 두개의 배열 A,B를 하나의정렬된 배열 C로 병합한다 해보자.     
각각 맨앞의 것을 단순하게 비교해 서 작은것을 배열 C에 넣으면 O(N)이 소요된다.(A+B의 원소수) 

```
A [X, , , , ] 
B [O,O,O,O, x ]   x 앞으로 다 c로 이동 

C [O,O ,O ,O ,O ,x    
```

그러지 않고 A의 맨앞의 원소의 B에서의 위치를 먼저 찾고  그 앞의 원소들을 다 C로 옮기면 log(N)이 소요됨.   
(x의 B에서의 위치 찾는데 이진 탐색 형식으로 logN걸림.)   

만약 이렇게 A먠앞 x원소의 B에서 위치 찾는데 그앞에 1개만 있다면? (즉 x포함 2개만 c로한번에 옮겨간다면?)   
log(N)비교 한번에 2개씩 만 옮겨간것.  
이것이 A+B 전체 원소수 N에 대해서 일괄적으로 항상 2개씩만 이동되면 전체 걸린시간은 (두개씩 옮겨가니까)   
N/2 x log(N)    
즉 아래 2가 즉 한번에 이동되는 수가 log(N)이하이면    
기존의 기초적인 병합과정에서 걸리던 O(N)보다  효율 떨어지는것. 


또한 비슷하게 A의 마지막 요소의 B에서의 위치를 찾는것    


```
A [X, , , ,y] 
B [O,O,O,O, x, , , ,y, *, *, *, ]   y뒤로 으로 다 c로 이동 

C [O,O ,O ,O ,O, x , y, * * * ]       
```

즉 순서를 나열 해보면   
1) 먼저 x의 b에서의 위치를 찾고     
2) b에서 그앞의 것들을 모두 c에 복사  
3) x 복사       
4) y의 b에서의 위치 찾고    
5) b에서 그 뒤의 것들 모두 c에 복사    
6) A에서 x,y사이에 있는것, B에서 x,y사이에있는것 적절히 순서대로 c에 복사    




**그렇다면 두방법.**  
**(기본적인 brute한방법, 한 배열 시작점,끝 점의 다른 배열에서의 위치 비교해서 일괄 복사하는방법)중 뭐가 나은지 어떻게 아는지?**   



기본적인 방법의 장점은 ?   
1) locality of reference : 각 배열에서 한개씩 꺼내 서 비교하므로 효율 좋음. 이미 캐시에 넣어놓고 하나씩 꺼내기도해서 더욱.    
2) simplicity : 필요한 변수의 수, 코드 복잡도    


한 배열 시작점,끝 점의 다른 배열에서의 위치 비교해서 일괄 복사하는방법 장점은?   
1) **potential** runs are large: 한번에 많은 비교 스킵가능함을 말하는듯.   
2) comparison are **potentially** reduce:    



둘중 어떤걸 쓸지 판단하는 방법은 -> **galloping(질주) threashhold**라는 파라메터로 판단.   

galloping threshhold는 7로 초기화 되어있다. -> 7개이상 한번에 가져올수 있어야 2번째 방법사용   
galloping은 B에서 요소들을 가져오는 과정이다.      
x가 b에 속하기 전가지 앞에서 뒤로 로 단순히 나아간다.     
또한 y가 b에 속하기 전까지 맨뒤에서 앞으로 나아간다.    
이런것을 gallop(질주)라한다.(기본적인 방법처럼 한개씩 비교,이동해나가는건 걸어가는것같아서 인듯)     

galloping이 계속 유효하게 효과적이면 galloping threashold 낮춰서 계속하기도함.  


예는 동영상을 참고하도록하자... 너무길어진다 ... 12분 56초 


`````python
# galloping고려 안하고 
# 다음 동영상 참고해서 짜봄.   
# https://www.youtube.com/watch?v=9kFHVe5MT6o 

def insertion_sort(a,start,end):
    
    # 삽입 정렬은 말그대로 오른쪽 미 정렬 부에서 꺼내서  왼쪽부분의 적절한 위치에 삽입 
    
    for i in range(start,end):
        for j in range(i,start,-1):
            if a[j-1] > a[j]:
                a[j-1],a[j] = a[j],a[j-1]
            else:
                break
        
    
#     print(a[start:end])
    return a[start:end]
    


def merge(a,start,end):
    
    mid = (start + end-1)//2       # 호출된곳 가보면  a[start:end]이런식으로 실제로는 end전까지만 가져가니까.
    i = start
    j = mid + 1 
    
    
    b =[]
    
    for k in range(start,end):
    
        
        # 잔 빈처리 
        if i > mid:
            b.append(a[j])
            j+=1
            
        elif j > end-1:
            b.append(a[i])
            i+=1
        
        elif  a[i]<=a[j]:
            b.append(a[i])
            i+=1
            
        elif a[i] > a[j]:
            b.append(a[j])
            j+=1
            

    return b






def tim_sort():
    
    # 먼저 5씩 insertionsort 수행   
    
    for i in range(0,len(a),5):
#         print(i)
#         print(a[i:i+5])
        print(i,i+5)
        a[i:i+5] = insertion_sort(a,i,i+5)
        print( a)
        print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
        
        
    # 5,10 단위로머지 하다가 크기 20 되면 머뭄   
    
    run_size = 5 
    while (run_size< len(a)):
        
        
        for i in range(0,len(a),run_size*2):
            a[i:i+run_size*2] = merge(a,i,i+run_size*2)
            
            print( a[i:i+run_size*2])
        
        
        
        run_size = run_size *2
        
    
    
    
    
    pass
    
    
# 전체 길이 20 이고, run크기가 5라고 하면 
# 5씩 끊어서 각각은 insertion sort를 수행하고  
# 이후에는 5+5 =10 합쳐서 두개 만들고 -> 10+ 10 합쳐서 20 큰거하나 로 만들어 합병함   

a =[54, 88, 77, 26, 93,      17, 49, 10, 17, 77,         11, 31, 22, 44, 17,          20,4,5,8,23] 
tim_sort()

len(a)
`````
