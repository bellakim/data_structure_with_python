# 6.1 해시테이블  

앞서배운 AVL, 래드블랙트리 의 삽입, 삭제 연산 수행시간--->O(logN)  
```
O(log) 보다 빠른 연산을 위해서  
키와 1차원 리스트의 인덱스의 관계 이용하여 키(항목)저장하자.  
```

### 내가 이해한 바를 정리하면
B트리 = 노드한개에 수많은 키 저장되어있어서 -->높이 낮춰서 탐색트리의 탐색 빠르게된다.  
       ( 즉 노드 안에서 들어온값을 수많은 키값들과 비교, 즉각적으로 분류--->자식노드에 연결 )  
B+트리 = 더 나아가서 리프에마나 유의미한 키값들 놓고 그위로는 root~리프 이르는 경로 로서의 의미지닌 노드들 존재   
**이것이 해시 테이블과 다른점은?**  
   B트리, B+트리의 높이가 1이라면 B트리,특히 B+트리는 해시테이블과 유사하지만    
   해시테이블은 인덱싱에 키값을(b,b+처럼 노드에 저장해서) 사용하지 않고    
   리스트의 인덱스 와 일치하는 키값을 해당인덱스의 리스트 공간에 저장한다.  
   (즉 해시테이블은 노드가 인덱싱에  동원되서 트리 높이 높아지면서 탐색시간 1보다 길어질 가능성 없다.) 

   만약 여기서 그대로 키값~인덱스값 일치하는곳에만 저장한다면      
   예를들어 1,999 이런 키 값이 들어왔을때 리스트의 1,999 인덱스에 해당하는 메모리만 차있고     
   나머지 공간은 None이여서 메모리 공간 낭비 + 키값을 탐색시 시간낭비 크다.    
   그래서 **해싱**(나눈 나머지중 일부만 취하는 등)을 통해서 비교적 범위 작은 **해쉬값**  
   만들어서 리스트의 인덱스로 이용한다.  
   항목이 이 해시값(인덱스)따라 저장되는 이 1차원 리스트를 **해시테이블**이라 한다.   
   (트리의 노드에서 키는 노드 정보의 위치위해 인덱스역할로 쓰이고 실제 정보는 value혹은 item이 담당했다.  
   해시테이블에서는 각각 키,정보 리스트에 담기는데 같은 해쉬값을 인덱스로 쓰는듯.)  
```                                    
                                        해시테이블 
                                      0  [   ]
          키 집합        해시함수     해시값  [   ]
         [... ,k]--->  h(k)=i-------> i  [ k ]
                                         [   ]
                                         [   ]
                                     M-1 [   ]
                                         
```

2개이상의 항목(키)가 해쉬함수 거쳐서 같은 해쉬값(인덱스) 나타내서 해시테이블(1차원리스트)의  
같은 원소에 저장해야될 일 발생한다 --->**충돌**(서로같은 키들 동일한 해시값 가질때)  

**충돌해결하는 두가지방법**  
1. 개방주소방식  
    충돌시 테이블의 빈 원소(리스트공간)에 충돌된 키(항목) 저장.
    (원래의 해시값과 다른곳에 저장되는것.-->어떤 규칙따라 저장할곳 탐색하고, 그 규칙따라 저장할때도 조회된다.)  
    
    
2. 패쇄주소방식  
    충돌이 일어나도 체이닝(해당 리스트 인덱스에 아예 리스트하나더 만들어버림) 등 이용해서 해당   
    해시에(리스트 인덱스) 충돌된 키 모두 저장

# 6.2 해시함수   
이상적 해쉬함수 = 균등하게(랜덤하게) 키들을 해쉬테이블의 인덱스(해쉬)로 변환시키는 함수  
(일상의 키들, 주민번호처럼(패턴있는) 유의미한 숫자 에서 앞자리등 가져오기 해버리면 충돌 많이 일어난다. 주번경우 맨앞뽑으면 남자끼리 다 충돌된다.)  

ex)  
1. **중간제곱함수**  
키^2 후 중간부분 취함.  
2. **접기함수**  
특정 자리수만 끊어쓰기.-->1000이면 15924에서 3자리 924만가져옴(어떤경우에도 0~999)  
3. **곱샘 함수**  
    1보다 작은 실수 를 키에 곱해서 정수만 해시값으로씀 (0.61803 성능좋다.)  

모든 자리수가 계산에 참여  
기존 키에 부여된 의미 알수없게된다.  

실제로는 해시테이블의 크기 M로 키 나눈 나머지를 해시로 많이 쓴다.  
(어떤경우에도 0~M-1의값.)

## 6.3.1 선형조사

처음발견한 empty원소에 충돌 일어난 키 저장.

```

```

리스트의 크기 M보다 i(초기해싱값) + j(탐색을위해 더한값) 커지면  
다시 0부터 시작하기위해 M으로 나눠 나머지를 취한거다.  

ex)  
M = 10, i = 7 이라면  
i+1, i+2 까지  리스트 a[i+@] 의 None체크하다가  
10들어오면  a[0]부터 empty체크, 
a[7] 초기값돌아와서 까지 empty없으면 저장실패.

## 6.3.2 이차조사  

충돌 후에 선형조사보다 멀리 떨어진 곳에서 empty 원소 찾는다.  
```
    (h(key) + j^2) % M, j=0,1,2,3

```  
이차조사는 바로 이웃하는(1건너의) 빈 곳이 채워져 만들어지는 1차 군집화 문제를 해결하지만  
같은 해시값을 갖는 서다른 키들인 동의어(synonym)들이 똑같은 점프 시퀸스를 따라 empty 원소 찾아 저장해서  
또다른 형태인 군집화인 **2차군집화**를 야기한다.  
또한 점프 크기가 1이 아니라 제곱만큼 커져서 리스트에 empty 원소 있는데도 건너 뛰어서 저장 실패할 수 있다.

````python
# 6-1 linear_prob
class QuadProbing:                
    def __init__(self,size):    #객체의 생성자   
        self.M = size           # 테이블크기 M
        self.a = [None]*size    # 헤시테이블 a
        self.d = [None]*size    # 데이터 저장용 d
        self.N = 0
                    
    def hash(self,key):     # 나눗셈 해시함수 
        return key%self.M 

    def put(self,key,data):
        initial_position = self.hash(key)
        i = initial_position
        j = 0 

        while True:
            if self.a[i] == None:
                self.a[i] = key
                self.d[i] = data
                self.N +=1
                return

            if self.a[i] == key : # data만 갱신  
                self.d[i] = data
                return
            j+=1
            i = (initial_position +j*j)%self.M

            
# 선형 조사에서는 다음과 같이 한바퀴돌면 break했다.            
#             if i == initial_position:
#                 break
            if self.N  >self.M:            #저장된 항복 수가 테이블크기보다 크면 저장 실패 
                break                      #(근데 빈곳 아니면 어짜피 저장 안하는데 M보다 클수있나?)
                                           # -> 아 그렇긴 하네. 그냥 단순히 꽉찬데 더 못넣는다는 뜻인듯.
                                           # N=M까지 간상황에서는 while문 돌필요도 없으니까.
                                         
                                           # 빈값 있는 상태에서 위조건 만족 못하고 break 안되고 
                                           # while무한일 수 있어서 최대 n회 수행 제한 거는것 좋음.  
    def get(self,key): # 탐색연산  
        initial_position = self.hash(key)
        i = initial_position
        j = 1
        while self.a[i]!=None:       # 탐색하다가 다음원소 None이면.. 그뒤로도 None이라는것.이때 정지
            if self.a[i] == key:     # 저장할때 그뒤로는 점핑해나가면서 저장 안해서 
                return self.d[i]     # 그뒤는 볼것도 없이 키를 해싱값+ 점핑해나가는 값 없을태니 
            i = (initial_position + j*j) % self.M
            j+=1

# 선형조사에서 이 조건은 1씩 증가했기에 100프로 한바퀴 돌면 다시 initail로 돌아와서 끝내면 됬지만 
# quadratic 이차 조사에서는 제곱한 값을 인덱스에 더해가며 탐색해서 다시 안돌아올 확률도 크고 다시돌아온다해도 
# 그 다음번이 처음다음의 다음이 아니므로 아래 조건은 알맞은 종료조건이 아니다.(한바퀴라는게 선형처럼없어서) 
#             if i == initial_position:
#                 return None
        
        return None
    
    
    def print_table(self): #해시테이블 출력  
        
        print('\n')
        for i in range(self.M):
            print(f'{i}', " ", end="")
        print('\n')
        
        for i in range(self.M):
            print(f'{self.a[i]}', " ", end="")
        print('\n')
        
        pass
````

````python
t = QuadProbing(13)
t.put(25,'apple')
t.put(37,'grape')
t.put(18,'pear')
t.put(55,'banana')
t.put(22,'lemon')
t.put(35,'orange')
t.put(50,'olive')
t.put(63,'watermelon')

print('탐색결과')
print('50의 data =', t.get(50))
print('63의 data =', t.get(63))

print('해시테이블:')
t.print_table()
````

이차조사의 put() 메소드 ---> 해시테이블에 empty있음에도 새항목 저장 못 할 수 있다.  
get()에서도 마찬가지. while루프 수행횟수 제한해 놓는게 좋다.

````python
# 6-6 chaing  
class Chaining:
    class Node:
        def __init__(self,key,data, link): # 노드 객체 생성자  
            self.key = key
            self.data = data
            self.next = link
        
    def __init__(self, size):
        self.a = [None]*size
        self.size = size
    
    def hash(self,key):
        return key%self.size
    
#     def put(self,key,data): #삽입연산           리스트 자체 넣는게 아니라 연결리스트의 맨 앞을 넣는듯. 
#         index = self.hash(key)
        
#         # 만약 리스트 해당 원소 비어있다면? 
#         # -->리스트 첫번째 노드로 저장-->다음 link는 None?
#         if self.a[index] == None:
#             self.a[index] = self.Node(key,data,None)
#         # 만약 어떤 원소가 있다면?
#         # 가장 맨앞에 놓고 원래 맨앞이었던 친구를 link로 둔다. 
#         else:
#             pre_first_node = self.a[index]
#             self.a[index] = self.Node(key,data,pre_first_node)


# 책보고 다시해봄. 
    def put(self,key,data): # 삽입연산 
        i = self.hash(key)
        p = self.a[i]
        
        while p != None:
            if key == p.key:         #이과정 거치는이유는 만약 data이미 갖고있는데 또저장하면 안되서(내가 빼먹은부분)
                p.data = data
                return 
            p = p.next  
        
#         a[i]가 None이던 아니던 next로 대입되야만 한다. 
        self.a[i] = self.Node(key, data, self.a[i])
        
        
    def get(self,key): # 탐색연산 
        index = self.hash(key)
        
        this_time_node = self.a[index]

        while this_time_node:
            if this_time_node.key == key:
                return this_time_node.data

            this_time_node = this_time_node.next

        return None

    
    
    def print_table(self): # 테이블 출력 
        
        for index, first_node in enumerate(self.a):
            print(index," ", end ="")
            
            this_time_node = first_node
            
            while this_time_node:
                print(f"-> [ {this_time_node.key}, {this_time_node.data}] ", end="")
                
                this_time_node = this_time_node.next
        
            print('\n')
    
    
    
        
        
    
        
````

```

이과정을 쫒겨난 키가 참조한 xtable(x(쫓겨난키)) 가 None이라서 바로 저장할 수 있을때 까지 계속한다.
```  

**아래는 책내용 이어서 정리**  

h(Key) ---> htable() 을 위한것 
d(key) ---> dtable() 을 위한것  

**알고리즘** 

```
[1] key = new_key                                       (new_key 새로 삽입된 키)(old_key 쫓겨나는 키) 
[2] h(key) = i 를 계산하여 htable(i)에 key를 저장한다.  
[3] if key가 저장된 원소가 비어있으면:
        삽입을 종료한다.  
[4] else: # key 저장되면서 그 자리에 있던 키를 쫓아낸 경우   
[5]        if old_key가 있었던 테이블이 htable이면  
                d(old_key)= j 계산하여 dtable[j]에 old_key 저장  
[6]        else: old_key가 있던 table = dtalbe이면  
                h(old_key) = j 계산하여 htable[j]에 old_key 저장  
[7] key = old_key, go_to_step[3]              
```

### 6.6 해시방법의 성능 비교 및 응용  

해시방법의 성능  
탐색,삽입연산 수행시 성공,실패 한 경우 각각 분석해 측정한다.  

**선형조사** 
적재율(차있는 capacity/ 빈 capacity인듯) a너무 작으면 해시테이블에 empty 원소가 너무 많고,  
a 값이 1.0에 근접할수록 군집화가 심화된다.  

**개방주소방식의 해싱**  
a(적재율)~ 0.5 즉 M(테이블크기) ~2N(키의 수) 반정도 차있을때  
O(1)시간 성능 보인다.  

**체이닝**
a가 너무 작으면 대부분의 연결리스트가 empty되고   
a가 너무 크면 연결리스트들의 길이 너무 길어져서 해시 성능 매우저하됨.   
M(테이블크기) 소수, a(적제율)10 (즉 각 테이블의 박스 길이 10정도의 리스트 이면) O(1)시간 성능 보인다.  

<img src="images/47376571_308316126456038_9008053205908586496_n.jpg" width="600">



**요약**    

1. 해싱이란 키를 간단한(해시)함수로 계산한 값을 1차원 리스트의 인데스로 이용하여 항목을 저장하고, 탐색, 삽입, 삭제 연산을 평균 O(1)시간 에 지원하는 자료구조이다.  
2. 해시함수는 키들을 균등하게 해시테이블의 인덱스로 변환하기 위해 의미가 부여되어 있는 키를 간단한 계산을 통해 '뒤죽박죽' 만든 후 해시테이블 크기에 맞도록 해시값을 계산한다. 대표적인 해시함수는 나눗셈 함수이다.  
3. 충돌을 해결하는 방법은 크개 두 가지로 분류되는데, 하나는 개방주소방식이고, 다른하나는 폐쇠주소방식이다.  
4. 개방주소방식에는 선형조사, 이차조사, 랜덤조사, 이중해싱이 있다.  
5. 폐쇠주소방식은 키에 대한 해시값에 대응되는 곳에만 키를 저장한다.  체이닝은 해시테이블 크기만큼의 연결리스트를 가지며, 키를 해시값에 대응되는 단순연결리스트에 저장하는 해시방식이다. 군집화 현상이 발생하지 않으며, 구현이 간결하여 실제로 가장 많이 활용되는 해시방법이다.  
6. 2방향 체이닝은 체이닝과 동일하나 2개의 해시함수를 이용하여 연결리스트의 길이가 짧은족에 새 키를 저장한다.  
7. 뻐꾸기 해싱은 뻐꾸기가 알버리는 모션 모방한 해싱방법, 탐색과 삭제를 O(1)시간에 보장하는 매우 효율적인 해싱 방법이다.  
8. 재해시는 삽입에 실패하거나 해시 성능이 급격히 저하되었을때, 해시테이블의 크기를 확장하고 새로운 해시함수를 사용해 모든 키들을 새로운 해시테이블에 저장하는것을 말한다.  
9. 동적해싱은 대용량 데이터베이스를 위한 해싱방법으로 재해시를 수행하지 않고 동적으로 해시테이블의 크기를 조절하는 방식이며, 대표적인 동적해싱에는 확장 해싱과 선형 해싱이 있다.
