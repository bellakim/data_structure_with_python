정의  
스레드 이진트리는 각 노드의 오른쪽 None 래퍼런스를 다음 방문할 노드를 참조하게 하고, 각 노드의 왼쪽 None레퍼런스를 직전 방문한 노드를 참조하게 한  이진트리이다.  
스레드 이진트리는 N+1개의 None래퍼런스를 활용하여 이전에 방문한 노드와 다음 방문할 노드를 가리키도록 만들어 순회 연산이 스택없이도 수행될 수 있도록 만든 리스트이다  
스레드 이진트리는 대부분의 경우 중위순회에 기반하여 구현된다.
스레드 이진트리는 스택을 사용하는 순회보다 빠르고 메모리 공간도 적게 차지한다는 장점을 갖지만 데이터의 삽입과 삭제가 잦은 경우 그 구현이 비교적 복잡해서 성능 안좋다.

또한 Node 객체에 2개의 Boolean 필드를 사용하여 레퍼런스가 스레드(다음방문할 노드를 가리키는)로 사용되는 것인지 아니면 left나 right가 트리의 부모 자식 사이의 레퍼런스인지를 각각 True와 False로 표시해주어야 한다

````python
# 코드로 한번 작성해보자. 


class Node:
    def __init__(self,item,left=None, right=None):
        self.item = item
        self.left = left
        self.right = right
        
class BinaryTree:
    def __init__(self): # 트리 생성자. 
        self.root =None

    def preorder(self,n): # 전위순회
        if n!=None: #맨 아래 level의 Node방문시 None이라서 함수 재귀호출 안하게 된다. 
            print(str(n.item),'',end='') # 맨먼저 노드 방문 
            if n.left:
                self.preorder(n.left)
            if n.right:
                self.preorder(n.right)
              
    def inorder(self,n): # 중위순회 
        if n!=None:
            if n.left:
                self.inorder(n.left)
            print(str(n.item),'',end='')
            if n.right:
                self.inorder(n.right)
    
    def postorder(self,n): # 후위순회 
        if n!=None:
            if n.left:
                self.postorder(n.left)
            if n.right:
                self.postorder(n.right)
            print(str(n.item),'',end='')

            
#문자열을 받아서 이진숫자로 치환해 보겠다.
def incoding_huffman(string1):

    alphabet_dict = {'A':0,'B':1,'C':2,'D':3}
    alphatbet_list = ['A','B','C','D']
    alphabet_rate_storage = [0]*4

    for i in string1:
        this_letter_index = alphabet_dict[i]
        # 해당 알파벳 등장시마다 빈도수 여기에 저장한다.
        alphabet_rate_storage[this_letter_index] = alphabet_rate_storage[this_letter_index] + 1
        
        
    #위 빈도수 바탕으로 튜플로 관리하겠다.
    # 튜블로 빈도수 : 알파벳 만들고 
    
    tuple_list = []
    for index,value in enumerate(alphabet_rate_storage):
        tuple_list.append((value, alphatbet_list[index]))
    
    
    tuple_list.sort(key=lambda x:x[0])
    tuple_list.reverse()
    print(tuple_list)
    
    
    def make_tree(tuple_list):
        
        #일단 각각의 노드 만들어 준다.  
        node_list = []
        for item in tuple_list:
            node_list.append(Node(item))
            
        count = 1
        while(True):
            
            if len(node_list) <2:
                break
            
            count +=1
            if count >30:
                break
                
            
            # 맨 뒤의 노드 두개를 뽑는다.
            node_1 = node_list.pop()
            node_2 = node_list.pop()
            
            # 작은거 두개 더해서 세로운 노드 만들고 그 좌우에 작으거 두개 놓는다.
            # 이 것의 알파벳 값은 없다.
            
            new_node_item = int(node_1.item[0]+ node_2.item[0])
            new_node = Node((new_node_item,0))
            new_node.right =  node_1
            new_node.left = node_2
            
            # 세로운 노드를 기존의 node_list에 넣고 item 순에 따라 정렬해 준다.  
            node_list.append(new_node)
            node_list.sort(key=lambda x:x.item[0])
            node_list.reverse()
            
#             print(node_list)
            
        
        
        # 이진트리 만들자. 
        
        tree = BinaryTree()
        tree.root = node_list[0]
        
        return tree

        
            
    binrary_tree = make_tree(tuple_list)
    
    
    # 이 트리를 순회하면서 가장 윗 트리에서 빈도수가 큰 
    # 트리에 0을 작은 트리에 1을 부여햐야 한다. 
    # root 제외 
    
    

    
    def give_binary_value(n,value): # 전위 순회 
    
        global result_tuple_list
        global binrary_tree
        
        if n!=None:
            print(str(n.item),'',end='')

            
            
            # 루트인 경우는 지나가 주자. 
            if n != binrary_tree.root:
                
#                 # item이 숫자인 경우  
#                 if type(n.item[1]) == int:
#                     value  = value +"0"
#                 #item 이 알파벳인 경우 
#                 else:
#                     value = value + "1"
#                     result_tuple_list.append((value,n.item[1]))
            
            

                result_tuple_list.append((value,n.item[1]))
            
            
            value_temp = value 
            if n.left:
                
                value = value + '0' 
                give_binary_value(n.left,value)
            
            if n.right:
                value = value_temp +'1'
                give_binary_value(n.right,value)

        
        return result_tuple_list
    
    value = ""
    #여기에 암호화 된 이진 수 다 들어있다. 
    result_tuple_list = give_binary_value(n=binrary_tree.root,value=value)
    
    
    # for문 돌면서 튜플의 두밴째 값이 문자가 아닌것들은 버리자. 
    #위의 튜플리스트를 알파벳을 키값으로 한 dict로 만들어 보자. 
    codded_dict = {}
    for value in result_tuple_list:
        
        if type(value[1]) == str:
            codded_dict[value[1]] = value[0]
            
            

    
    
    
    
#     # 초기에 받은 string1을 문자 -> 이진수로 치환해 보자.  
    
    result_huffman_string = ""
    
    
    for value in string1:
        
        binary_value = codded_dict[value]
        result_huffman_string  = result_huffman_string + binary_value
        
    
    
    
    return result_huffman_string
            
            

result_tuple_list =[]
result_huffman_string = incoding_huffman("CDDCACBCBCCCBBCDA")
# binrary_tree.root.item
# binrary_tree.inorder(binrary_tree.root)
print('')
print(result_huffman_string)
````
