 # 1. 선택정렬 

```python
# 선택정렬 다시 짜봄.    

# 선택정렬은 배열이있으면 그중 최솟값을 한개씩 빼서 왼쪽부터 끝까지 정렬시켜 나가는 방식. 



# 1. 손코딩  
def selection_sort(a):
    
    for i in range(len(a) -1 ):
        
        min_idx = i 
        for j in range(i,len(a)):                
            if a[j] <  a[min_idx]:
                min_idx = j
        a[i],a[min_idx] = a[min_idx],a[i]
            
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
selection_sort(a)
print('정렬 후:\t', end="")  
print(a)  

# 2. 시간복잡도 최선,평균,최악.
#    최선: 모두 처음부터 정렬되 있다 해도, 결국에는 매번 1,2,... n-1번 비교 -> n(n-1)/2   
#    최악,평균의 경우도 같음. 
#    다만 이동이 최선 -> 0번, 평균, -> n/2번? 최악 -> n/2번   

# 3. stable? inplace? 
#    위에 짠 대로라면 만약  1a, 1b 이렇게 나열되 있더라도 최솟값이 1a의 인덱스가 먼저 자리해서  1b< 1a 성립 안하기 때문에  
#    1b는 그다음 회차에 최솟값으로서 나가게됨 -> stable 하다 -> 아니다 이생각 틀리다 !!!!!!!!!!!
#    
#    만약 43341 이렇게 나열된 경우 초기 min_idx가 0자리라서 마지막 최솟값인 1과바뀌면 13344 로  4끼리 위치가 바뀜 -> unstable
#
#    추가 메모리 필요 x 

# 4. 장점, 용도.  
#    비교적 구현 간단해서 코드 줄이 적다  -> 코드 메모리가 중요한 곳에서 쓰이나?   
#    버블보다 값의 이동이 적음   
# 
```

# 2. 삽입정렬 

```python
# 삽입 정렬 다시 짜봄.

# 삽입정렬은 왼쪽에 정렬된 리스트를 축적해 나가는 방식인데     
# 오른쪽 정렬되지 않은 부분에서 값을 순서대로 하나씩 끊어와서 왼쪽 정렬된 리스트에서 적절한 위체에 넣어주는것.   
# 뽑은위치 -1 부터 맨 앞까지 가면서 한칸씩 비교해 나가 뽑은위치의 적절한 위치를 찾는다. 
# 

# 1. 손코딩  
def insertion_sort(a):
    for i in range(0,len(a)):
        for j in range(i,0,-1):               # 이게 -1이 아니라 0까지인 이유는 아래서 j, j-1 비교시 
            if a[j] <  a[j-1]:                # -1이라면 0, -1 즉 마지막것을 비교하게되서 꼬이게됨. 
                a[j],a[j-1] = a[j-1],a[j]     # 0 이라면 1 ,0을 비교하니 괜찮다. 
            else:
                break 
    
    
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
insertion_sort(a)
print('정렬 후:\t', end="")  
print(a)

# 2. 시간복잡도 최선,평균,최악.
#    최악의 경우 : 아예 역순인경우  ->   9,8,7,6,5,4,3,2,1  매i마다 맨앞까지 타고 타고 교환해줘야함. 비교,교환 모두 1,2,....n-1 번 즉 O(N^2)
#    최선의 경우 : 처음부터 정렬인 경우 - > 1,2,3,4,5,6..  매 i마다 1번씩만 비교함 ->  비교 n-1, 교환 0  O(N)
#    평균의 경우 : 매 i마다 중간까지 만 교환하는경우. 비교, 교환 모두 n(n-1)/2 -> O(N^2)

# 3. stable? inplace? 
#   812385323 이런식으로 있다면..앞의 것이 교환되는 과정에서 순차적으 타고 타고 어쨋든 간다. -> stable  
#   inplace다.   

# 4. 장점, 용도.    
#    장점 : 이미 정렬 되있는것에 신규값 추가해서 정렬하는경우(삽입경우)성능좋음 / 안정적이고, 추가메모리x인것도 장점./ 입력크기 작은경우 우수   
#    단점 : 입력이 역으로 정렬되있으면 매우느림 O(N^2) -> 이경우 이동을 덜하는 선택정렬이 효율더 좋음 / 입력에 따라 수행시간 민감하게 바뀌는편. 
#    용도 : 머징 소트 에서 부분의 n이 작을때 대신 삽입정렬 쓰기도함. -> 팀소트 (n작을때는 합병보다 효율 좋다.)
```

# 3. 버블정렬 

```python
# 버블 정렬 다시 짜봄.
# 버블 정렬은 앞뒤로 한 요소씩 비교해 나가면서 더 큰것이 앞에 있으면 서로 교환해주는형태    
# 이것을 처음에는 1자리 ~ n 자리 까지 비교  -> 1자리 -> n-1자리 비교   이런식으로 1자리 -> 2자리 까지 비교 해나가는 형태    
# 맨 뒤부터 큰값이 쌓인다. 

# 1. 손코딩  

def bubble_sort(a):
    for i in range(0,len(a)):
        for j in range(0,len(a)-i-1):
            if a[j] > a[j+1]:
                a[j],a[j+1] = a[j+1],a[j]
                
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
bubble_sort(a)
print('정렬 후:\t', end="")  
print(a) 


# 2. 시간복잡도 최선,평균,최악.
# 최선,최악,평균 모두 결국에는 각 i 에서 n-1, n-2, n-3 ... 1번 비교,  이동은 각각 0,N^2,N^2인듯 . 시간복잡도는 즉 O(N^2) 

# 3. stable? inplace? 
#    stable한지? : 8,1,2,3,4,8,6,7이런식으로 배열시  
#                 맨처음 i 번에서 8이 뒤로 가다가    1,2,3,4,8,8,6,7 이순간이 오는데  
#                 이때 8이 서로 안바뀌기 때문에 뒤의 팔이 먼저 맨 뒤로 가서 정렬완료 되어버린다.  -> stable 하다. 

# 4. 장점, 용도.  
#    장점.. 안정적인것, 수행시간 거의 동일한것. 
#    용도 : 다른 정렬 수행후 완료 됬는지 판단 할때 씀. (앞뒤 하나씩 비교해 나가다가 하나라도 틀리면 False리턴하는 방식)    
#    
```

# 4. 쉘정렬 

```python
# 쉘정렬 다시짜보기 (다시보도록) 

# 쉘정렬은 삽입정렬과 비슷한데 좀 다르다. 
# 일정 간격을 정해놓고 그 간격을 점프한 대상들끼리 삽입정렬을 수행한다보면됨 
# 이것을 한번 수행한 결과를 보면 배열에서 앞에있는 원소들이 얼추 작은 값들이 많다.  
# 이것은 삽입정렬 하려 할때  배열이 역으로 되었거나, 전체적으로 이동할것이 많을때 전처리를 해줘서 대략적으로라도 순서 맞게 먼저 해주는 기능이라 봐도될듯.    
# 일정 간격은 특정 규칙에의해 매번 줄어들고 결국에 1이되면 일반적인 삽입정렬과 완전히 같게된다.    

# 좀더 자새히 설명하면 매 h마다 h+1에서 for문을 시작하고     
# a[i - h] >a[h] 이면 값을 교환해 준다. -> 맨 앞까지 삽입때 처럼 밀어줌.(while문) 

# 1. 손코딩  
def shell_sort(a):
    
    h = 4            # 3x+1간격: 1, 4, 13,40, 121...중에서 4와 1만 사용.   

    while h >=1:
        for i in range(h,len(a)):
            j = i 
            while(j>=h and a[j] < a[j-h]):
                a[j],a[j-h] = a[j-h],a[j]
                j = j-h 

        h  = h //3 
                
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
shell_sort(a)
print('정렬 후:\t', end="")  
print(a) 

# 2. 시간복잡도 최선,평균,최악.
#     간격설정 따라 다르고... 복잡아주 복잡하다.  최적 간격을 찾도록하자.
#     대략 찾아본 바로는 최선 n / 평균 n^1.5 / 최악 n^2 

# 3. stable? inplace? 
#    stable : 예상했겠지만 앞뒤로 하나씩 이동이 아니라..안정성이 보장 안됨.  
#             예를 들어  712345718313       이렇게있을때 간격 설정 할때 두 7이 포함되게 안한다면 순서 바뀔 수도 있는것. 
#    inplace다 

# 4. 장점,단점,용도.  
#    장점 : 기존 삽입정렬에서 값들 한칸씩 이동하는것을 보완/ 전처리과정 으로 작은값 리스트의앞으로 ~ 큰값 리스트의 뒤로 위치시키게함 
#          입력 작은경우 성능 좋음 ? 
#          삽입, 선택 정렬 비해서도 큰 n 에 대해서 빠르게 정렬함.   
#    단점 : 안정성없다. 
#    용도 : 
#         간격따른 그룹별 정렬 알고리즘을 하드웨어 설계 에서쉬워 -> 임베디드에 사용 
```

# 5. 힙정렬  

```python
# 힙소트 다시 해보기 (또보기...)    

# 힙소트는 주어진 배열을 가지고 상향식 힙을 만들고   
# 힙 속성에 의해 맨위에 위치하게된 최대값을 힙의 맨 뒤에 넣어주는 방식으로 정렬함.  
# 초기의 햅 생성에서는 leaf노드 바로위의 부모 단 부터 ~ 0 번 인덱스 까지 다운힙을 수행하는방식으로 힙속성을 회복하고   
# 힙소트 과정에서는 맨위의  root를 빼서 맨뒤와 swap후에 맨위로간 것 부터 down힙을 수행해서 힙속성 만족하는 곳까지 내려간다.  
# 이것을 반복하는데 반복할때마다 swap된것을 다운힙 하는 인덱스가 1씩 줄어든다. (뒤부터 큰값 쌓이니까 그부분은 비교 x인것 )    

# 필요한 함수는 1. 다운힙  2. 힙생성  3. 힙소트 

# 1. 손코딩  


def downheap(i, size): # 루트로 올라온 키에 대해 힙속성을 회복시킴   
    while 2*i <=size:                                             
        k = 2*i
        # 좌 우중 승자 결정~승자하고 비교하겠다.
        if k < size and a[k] < a[k+1]:                            # k < size 라서 여기 못통과하면 and 이하 대입안함 
            k+=1
        if a[i] >= a[k]: # 힙속성 만족하면 루프 나간다.                  # 힙속성 만족시 바로 나가도록...주의! =인것 주의 !  
            break                                                  # (걍 끝까지 해도 되긴하지만.. 효율상 ! ) 
            
        a[i], a[k] = a[k], a[i] # 자식원소가 더 크면 바꿔준다.
        
        i = k
        
        
def create_heap(a): # 정렬하기 전 최대 힙 만들기.
    # 리프 노드의 부모 단 ~ 0 까지 다운힙 수행   
    
    size = len(a)-1
    
    for i in reversed(range(1,size//2+1)):                 # 2m, 2m+1을 자식으로 놓기 위해 1부터 시작함을 주의 
        downheap(i,size)
    

def heap_sort(a):
    N = len(a) - 1
    for i in range(N):
        a[1],a[N] = a[N],a[1]  # 루트와 히ㅃ의 마찌맊 항몪 교환
        downheap(1,N-1)   # 왜 -1이여야 하지 이미 위에서 len(a)-1해줬는데? ->아..마지막은 교환되서 최대값이 맨뒤에 N번째에 있는것. 
        N -= 1             # 힙 크기 1 감소  
                           #(이걸 해야 뒷부분은 추후 뒷부분
                           # <최대값부터 순차로 저장된부분>은 다운힙시 안건든다.)
        

a = [-1,54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]  
print('정렬 전:\t', end="")
print(a)
create_heap(a)
print('최대힙:\t', end="")
print(a)

heap_sort(a)
print('정렬 후:\t', end="")  
print(a) 

# 2. 시간복잡도 최선,평균,최악.
#   평균,최선,최악 모두 같음 -> nlogn 인데 상향힙 구성에 O(n) + n번 다운힙을 높이 만큼 해서 nlogn

# 3. stable? inplace? 
#    stable ->  상향힙 구성 과정 살펴보면... 2m ~ 0 까지 다운힙시 부모>= 자식 일때 교환 안하고 다음 다운힙으로 넘어가는지라 안정  
#               힙소트 함수 내에서는 차례로 root를 꺼내 뒤로 붙이는데 17a,뒤에  17b 있다면 ? 17a가 먼저 뒤에 붙고 그 앞에
#               17b가 붙어 원래 순서 틀어지는듯  -> unstable    
#    inplace맞다 . 

# 4. 장점,단점.  용도. 
#    장점:  메모리x/ 어떤 입력에도 nlogn/ 
#          계속 데이터가 추가되거나 빠질 때, 그 순간마다 정렬된 상태를 유지할 수 있다는 장점  
#          recursion없이 작성 가능하고, 이 경우에 추가적인 메모리를 필요로 하지도 않습 
# 
#    단점 : 루프내 긴코드, 비효율적 케시메모리사용 -> 대용량 소팅에부적절함. 
#    용도 : 가장 유용히 쓸때는 큰값 몇개만 필요할때. 

```

```python
# 합병정렬 다시 해보기    

# 1. 손코딩  
def merge(a,b,low,mid,high):  
    
    i = low           # 앞부분 비교시작점 
    j = mid + 1       # 뒷부분 비교 시작점  
    
    for k in range(low,high+1):                #k = 저장 시작점. 
        #잔반처리 
        if i > mid:      # i 가 미드까지 올라왔다면  
            b[k] = a[j]     # 그이상 올라간다면 j 의시작점.
            j+=1            # 걍 j 증가시키고 저장해주면됨 
                
        elif j > high:
            b[k] = a[i]
            i+=1
        
        
        elif a[i] <= a[j]:
            b[k] = a[i]
            i+=1

        elif a[i] > a[j]:
            b[k] = a[j]
            j+=1
            
    # b를 low부터 high 까지 a에게 최신화 시켜주자.  
    for k in range(low,high+1):
        a[k] = b[k]
    

def merge_sort(a,b,low,high):  
    
    if low >= high: # 원소 1개 이하 정렬시키려하면 걍 리턴함.
        print(low,high)
        return      # 우리는 a에 정렬이 바로 반영되므로(b거치긴하지만) 
                    # 정렬되있다면(원소한개라 정렬필요없다면) 그냥 리턴만 해도됨. 
        
    mid = (low+high)//2
    merge_sort(a,b,low,mid)      # 앞부분 정렬 시킴 
    merge_sort(a,b,mid+1,high)   # 뒷부분 정렬 시킴 
    merge(a,b,low,mid,high)      # 앞 뒷부분 합병
                                 # (두 시작점 i,j부터 각각 -> mid, high까지 타고가면서                      
                                 # 더작은 것을 b에저장후 마지막에 a에 반영함.                      
            
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]
b = [None]*len(a)
print('정렬 전:\t', end="")
print(a)
merge_sort(a,b,0,len(a)-1)
print('정렬 후:\t', end="")  
print(a) 

# 2. 시간복잡도 최선,평균,최악.
#  최선,평균,최악 모두 nlogn 
#  T(N) 이 N개 정렬하는데 걸리는 시간이라면 T(N) = CN + 2*T(N/2)
#                                         = CN + 2*(CN/2 + 2*T(N/4)) = 4*T(N/4) + 2CN 
#                                         = 2CN + 4*(CN/4 + 2*T(N/8)) = 8*T(N/8) + 3CN  = 2^K * T(N/2^K) + KCN 
#                                        N =2^K 되는순간의 k = logN   -> NT(1) + NClogN = O(NlogN)  

# 혹은 각 제귀 단계에서 총 n번의 비교를 하는데 각 단계가 log n 번의 깊이 이기 때문에 -> O(NLOGN)   

# 3. stable? inplace? 
#    stable? :  합병하는 과정에서 보면 
#               만약 같은 값을 비교 했다면 a[i] <= b[j] 이면  b[k] = a[i]를 넣는데 순서상 i->j 이기에 i가 먼저 들어가므로 stable맞다.
#              
#               다른경우에  a[i] = a[j]이지만  j 뒤에 i 오도록 정렬되는 경우는? 
#               만약 [3,4,5,1a] [1b,5,6,7]  이것을 합병한다면 뒤 집합의 1이 앞에 오게 되고 앞집합의 1이 뒤에오게 될것 
#               하지만 제귀한다 했을때 보자  
#               [3,4][1a,5], [1b,5],[6,7]
#               [1a,3,4,5,] [1b,5,6,7] 
#               [1a,1b,3,4,5,5,6,7]       -> 재귀 들어가서 길이 1개일때부터 고려하면 stable맞는듯! 
#               
#
#   inplace아니다. (inplace머지소트있긴한데 잘 안쓰는듯..복잡해서?)   


# 4. 장점,단점.  용도. 
#    장점: 모든경우 nlogn, 안정성 높다 . 
#    단점: 추가 메모리 사용. 
#    용도: 외부 디스크 정렬에서 임의접근 비싸서 못하고, 순차접근이 용이할때 쓰이는듯. 
#         자료구조 연결리스트로 구성되있으면 병합 정렬중 연결들만 바꾸면 되서 퀵보다 빠름.  
```

# 7. 퀵소트 

```python
# 퀵정렬 다시 해보기    

# 퀵소트 재귀적으로 동작하며 합병정렬과 비슷하나  합병은 먼저 재귀로 끝까지들어가서 합칠때 값비교후 큰 정렬된 덩어리를 만들어 나가지만   
# 퀵소트는 먼저 값을 비교한후 쪼개 어나가면서 값을 비교해서 정렬해나간다. 또한 일정 크기로 나뉘지 않고피벗을 기준으로 그룹을 나눈다. 
# 피벗을 어떻게 정하느냐에 따라서 효율이 많이 달라진다. 
# 동작하는 방식은 i,j를 배열 양끝에서 중앙으로 달리게 해서 각 그룹의 조건을 만족하지 않는곳에서 멈춰서 서로 교환한다. 
# 필요한 함수 1. quicksort, 2.partition 


# 1. 손코딩  
def qsort(start,end):
    
    if start >= end:
        return

    pivot = partition(start,end)
    
    qsort(start,pivot-1)
    qsort(pivot+1,end)

def partition(start,end):
    
    pivot = start 
    
    i = start + 1 
    j = end 
    
    while True:
        
        # 그룹 속성 만족하는 동안 가운대로 진행시킴 
        while i < end and a[i] < a[pivot]:
            i+=1
        
        while j > start+1 and a[j] > a[pivot]:
            j-=1
            
        if i >=j:
            break
            
        a[i],a[j] = a[j],a[i]
        i+=1
        j-=1
        
        
    a[j],a[pivot] = a[pivot],a[j]
    
    
    
    return  j
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]

print('정렬 전:\t', end="")
print(a)
qsort(0,len(a)-1)
print('정렬 후:\t', end="")  
print(a) 

# 2. 시간복잡도 최선,평균,최악.
#    최선의경우 nlogn이다. 
#    피벗에 의해 메번 정확히 2등분으로 나뉘는 경우 인데 각 제귀에서 n번의 비교를 하고 
#    n개를 정렬시 제귀를 총 logn 번 하기 때문에 nlogn이다. (T(N) = cN + 2*T(N/2) 로 증명하는 방법도 같음. -> 평균의경우 이렇게증명 )    
#    
#    최악의 경우 역순으로 정렬됬을때 매번 가장작은값, 가장 큰값을 피벗으로 고르는 경우이다. 
#    이경우 T(N)  = N-1 +T(N-1)
#                = N-1 + N-2 + N-3 +......T(1) = O(N^2) 

# 3. 성능향상법? 
#    1. 피벗을 고를때 무작위 3개 or 9개의 평균과 가까운 값을 피벗으로 정해서 최대한 그룹을 균등하게 둘로 나눈다.   
#    2. 합병성능향상 때처럼 일정 크기 이하에서 삽입정렬 수행 (작을경우 삽입이 효율 좋음)   
#    3. 랜덤 섞기를 처음 한번 한다.      

# 4. stable? inplace? 
#    stable하지 않다. ->  퀵소트 시에 피벗을 기준으로 그룹이 갈리는데 만약 피벗 값이 비교되는 값과 같은 경우   
#                       ior j 값이 멈춘곳과 나머지에서 멈춘값이 교환된다. 
#                       만약 피벗이 7이고     1,2,3,4,7a,7b,5,4,3  이것을  각그룹 기준 a <7, b> 7 맞을때까지 중앙으로 i,j를 전진시킨다면   
#                       
#                                                 
#    inplace이다.  


# 5. 장점,단점.  용도. 
#    장점: 일반적인 경우 가장 빠름. 
#         가상메모리 (지역적)적중률 높아서 가상메모리에서도 빠름.(피벗 중심 값교환) 
#    단점: 안정하지 않고, 최악의 경우 n^2 / 거의 정렬된 데이터 경우 버블, 삽입이 더 빠르기도함. 
#    용도: 무작위 데이터에 대한 일반적인 경우 사용. 
```

# 퀵소트 vs 합병정렬    
**퀵소트**가 일반적인 경우 가장빠름.
 하지만 nlogn을 무조건 보장하지 않고, stable하지 않고, inmemory아님.  
(pivot을 중간값으로 먼저구하는 과정 포함하면 nlogn이지만 이경우 그냥 하는 nlogn보다 느려져서 그냥 얼추 pivot 을 3개중 중간값 이런식으로 구하는듯  
. 최악의경우 감소하고라도)  
이경우 에 모든경우 nlogn인  **힙소트**가보통  대체제로 쓰인다.  
하지만 이경우 상수계수 c가 더 크고 소요시간의 상한이 좀더 분명해야할때만 쓰이는듯.  
**합병정렬**은 자료구조가 임의접근(random access -기억장치에 저장된 자료에 순차접근안하고바로접근)  을 제공안할때 쓰인다. 동작하는 구조를 보면 >순차접근 이라 그런듯.  
또한 임의접근이 매우 비싼 외부 정렬(디스크나,테이프에서) 에서도 쓰인다.  
(보통 메모리에 정렬하려는 파일 전체 안들어가면 청크로 짤라서 메모리에 넘겨 퀵소트 하고 외부저장소에서 는 그것 받아서 병합정렬 수행해 정렬된 파일
 만드는듯)  

**정리하면 퀵소트 말고 합병절렬 쓰는경우는**  
1. 연결리스트에 데이터가 담겼을때  
2. 최악의 경우 O(N^2)피하고싶을때  
3. 안정적인 정렬 필요할때   
4. 큰데이터가 하드드라이브에있고 결과를 하드에 저장해야할때   
5. 비교함수가 느려서 비교횟수를 최소화하고 싶을때

# 8. 기수정렬 

```python
# 기수정렬은 키가 여러개 일때 각 키를 기준으로 정렬하는것인데 
# 앞부터 인지 뒤부터인즈 ,lsd, msd 기수정렬로 나뉜다. 
# 여기서 해볼것은 lsd 이다. 
# 예를들어 ABC , ABE, CDA 이것을 가장 작은 3번째 자리기준으로정렬 -> 2번째기준정렬  ->1번째기준정렬 
#  이렇게 순차적으로 정렬하는것.   

# 1. 손코딩 

# 숫자 포함하는 모든 문자 (아스키코드)에 대해 길이가 3인 스트링들을 사전적  순서로 정렬  
def lsd_sort(a):
    WIDTH = 3       # 입력:3개의 문자로 된 스트링   
    N = len(a)
    R = 128         # 문자 종류 수(UTF-8, ASCII)  
    
    
    for d in reversed(range(WIDTH)):
        
        temp = [None]*N
        count = [0]*(R+1) # 아래 빈도수 계산에서 1씩 밀어서 저장하려고 이렇게 할당함 
        # 1. 빈도수 계산
        for i in range(N):
            count[ord(a[i][d])+1]+=1
        
#         print(count)
        # 2. temp에 저장할 시작 index계산    조사한 빈도를처음부터 더해 나가서 각 기의 시작 인덱스 를 구한다.  
        for j in range(R+1):
            count[j] = count[j-1]+count[j]
#         print(count)
        
        # 3. 위치 최신화 및 temp에 저장          실제 행렬 a 를 구한 인덱스 기준으로 정렬한다.
        for k in range(N):
            p = ord(a[k][d])
            temp[count[p]] = a[k]
            count[p] +=1
            
#         print(temp)
                
        # 4. a에 복사 
        
        for i in range(N):
            a[i] = temp[i]
            
        print(a)
        
a = ['ICN','SFO','LAX','FRA','SIN','ROM','HKG','TLV','SYD','MEX','LHR','NRT','JFK','PEK','BER','MOW']
print(a)
lsd_sort(a)
print(a)

# 2. 시간복잡도 (최선,평균,최악)  
#    최선,평균,최악 =   d(키자리수)* (N(입력크기) + R(기수가짓수) = d(N+R)

# 3. stable? inplace?  
#    stable?  :  342a      -> 342a   -> 113
#                232b         232b      132b     stabe아니다. 
#                113          113       342a   
#    inplace아니다.

# 4. 장점? 단점? 용도?
#    장점 :  제한된 범위내 숫자 (주민번호, 전화 번호와 같이 radix의 자릿수, 가짓수 고정되있는경우.) 효율적임 
#    단점 :  일반적인 경우에 대해서 사용 못함.(위같은경우는 알파벳 에 용도 고정) / radix , temp등 추가 메모리 공간 필요 
#           (힙정렬의 단점처럼) 크기 커질수록 케시메모리 비 효율적으로 씀 / 루프내 코드 너무많음. 
# 
#    용도 :  제한된 범위내 숫자 (주민번호, 전화 번호와 같이 radix의 자릿수, 가짓수 고정되있는경우.)  
```

```python
# 1. 손코딩 
def insertion_sort(a,start,end):
    
    # 삽입 정렬은 말그대로 오른쪽 미 정렬 부에서 꺼내서  왼쪽부분의 적절한 위치에 삽입 
    
    for i in range(start,end):
        for j in range(i,start,-1):
            if a[j-1] > a[j]:
                a[j-1],a[j] = a[j],a[j-1]
            else:
                break
        
    
#     print(a[start:end])
    return a[start:end]
    


def merge(a,start,end):
    
    mid = (start + end-1)//2       # 호출된곳 가보면  a[start:end]이런식으로 실제로는 end전까지만 가져가니까.
    i = start
    j = mid + 1 
    
    
    b =[]
    
    for k in range(start,end):
    
        
        # 잔 빈처리 
        if i > mid:
            b.append(a[j])
            j+=1
            
        elif j > end-1:
            b.append(a[i])
            i+=1
        
        elif  a[i]<=a[j]:
            b.append(a[i])
            i+=1
            
        elif a[i] > a[j]:
            b.append(a[j])
            j+=1
    return b

def tim_sort():
    
    # 먼저 5씩 insertionsort 수행   
    
    for i in range(0,len(a),5):
#         print(i)
#         print(a[i:i+5])
        print(i,i+5)
        a[i:i+5] = insertion_sort(a,i,i+5)
        print( a)
        print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
        
        
    # 5,10 단위로머지 하다가 크기 20 되면 머뭄   
    
    run_size = 5 
    while (run_size< len(a)):
        
        for i in range(0,len(a),run_size*2):
            a[i:i+run_size*2] = merge(a,i,i+run_size*2)
            
            print( a[i:i+run_size*2])
        
        run_size = run_size *2
        
        
    
# 전체 길이 20 이고, run크기가 5라고 하면 
# 5씩 끊어서 각각은 insertion sort를 수행하고  
# 이후에는 5+5 =10 합쳐서 두개 만들고 -> 10+ 10 합쳐서 20 큰거하나 로 만들어 합병함   

a =[54, 88, 77, 26, 93,      17, 49, 10, 17, 77,         11, 31, 22, 44, 17,          20,4,5,8,23] 
tim_sort()

len(a)


# 2. 시간복잡도 (최선,평균,최악)  
# 일반적인 경우 
#  머지 의한 Cmlog(n-x) + 삽입정렬에 의한 이동 O(C^2)  -> 작은 x 빼줘서 전체 병합 트리 높이 낮추고, 삽입정렬 작은 x만큼해서 시간소요줄어듬   


# 3. stable? inplace?  
#    삽입, 병합 둘다 stable이라서 stable한듯 

#    inplace아니다. 

# 4. 장점? 단점? 용도?  
#    장점: 실제 세계의 값들이 부분적으로 정렬되어 있을때 팀소트 유용.
#         최악의 경우 nlonn이라서 퀵소트처럼 최악의경우 n^2에 빠지지않고,최선의경우는 n 이라서 힙,머지소트 보다 최선의경우도 더빠르다.  
#         (합병의 최악의 경우, 삽입의 최선의 경우 시간을 가져가는것.)
#          소요시간 빠른 세가지 이유 1. 작은 단위에선 삽입정렬 수행 해서  
#                               2. 스택 이용해서 비슷한 크기의 청크끼리 병합 -> 합병효율 좋다 
#                               3. galloping(질주)모드 이용해서 한집합 을 다른집합에서의 위치 찾아 일괄적으로 값 이동시켜버림 
# 
```
