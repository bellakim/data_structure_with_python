Amortized Analysis 에서 수행시간을 분석하는 방법들 중  accounting method를 알아보겠다.  
**accounting method** 란 연속적인 연산들이 최악의 경우를 만들 때, 그때의 한번의 연산에 대한 "평균비용"을 구하는 방법이다.(어떤 연산을 수행하려면 지불해야하는 비용이 있다.)  

accounting mehtod는 이렇게 정의된다.  
amortized cost(평균비용) = actual cost + accounting cost  

accounting method 는 0보다 커야한다.  
actual cost (stack의 push를 예로들면 push 하는데 드는 비용)  
accounting cost (우리가 가지고있는 남아있는 비용)  

정리하면 accounting method는 어떤 연산에 대해 비용을 지불하는데, 그 비용은 실제 일하는데 쓰이는 비용 + 적립하는 비용 = amortized cost라 한다. 이게 얼마인지 알아내는 것이 accounting method의 목적이다.

Doubling이 일어나는 순간 1 + tn 의 비용이 필요하다.  
이 비용은 계좌에서 가져오는것인데  
Doubling이 일어나는 순간 내 계좌에 2t가 적립되면서 1+tn이 빠져나간다.(여기서 1은 무시되는듯.)  

정리하면 Doubling이 일어났을 때의 push 연산에 대한 accounting cost는 -txn+2t이다.

amorized cost = actual cost + accounting cost  
Doubling 안일어 났을경우 1+2t  
Doubling 일어났을 경우 (1+tn) + (2t - txn) = 1+ 2t   

즉 Array Doubling을 이용해 구현한 Stack의 push 연산에 대한 복잡도는 amortized O(1+2t)

분할상환분석

분할상환분석 (amortized analysis)은 주어진 알고리즘의 시간 복잡도나 프로그램을 수행하는데 소요되는 시간 또는 메모리 같은 자원 사용량을 분석하기 위해서 사용하는 기법이다.알고리즘을 분석할 때에 각각의 연산마다 최악의 경우를 따져본다는 것은 굉장히 힘든 일이기 때문에, 이를 쉽게 해결하기 위해 분할상환분석이라는 방법론이 나오게 되었다.

어떠한 임의의 알고리즘에 대해서, 어떤 연산은 자원적 측면에서 상당한 비용을 소모할 수 있지만, 반면 다른 연산은 그렇게 고비용을 소모하지 않을 수 있다. 분할상환 분석은 알고리즘의 전반적인 연산 집합에 대해 비용이 높은 연산, 그리고 비용이 덜한 연산 모두를 함께 고려하는 기법이라 하겠다. 이것은 다른 종류의 입력, 입력의 길이, 이 알고리즘의 성능에 영향을 미치는 다른 요인들을 전부 고려한다. 수행된 모든 연산에 대해 자료구조 연산만의 어떤 시퀀스를 수행하는데 필요한 시간의 평균을 구한다. 비록 그 시퀀스에서 하나의 연산비용이 비싸더라도, 그 일련의 연산에 대해 평균을 구하면 연산 하나의 평균 비용이 작다는 것을 분할 상환 분석을 이용해 보일 수 있다.

**o(small-oh)표기법**:  
f(x) = o(g(x))와 같은 빅오 표기법이 x가 무한대로 커질때 뿐만 아니라 특정 값에 가까워 질때에 대해서도 사용될수 있는 개념 인 반면 small-oh는 x가 무한대로 커지는 상황에서만 사용되는 개념이고 Big O보다 더욱 엄격하다. (. Big-O는 포괄적 인 상한선이며 little-o는 엄격한 상한선입니다) **g(x)의 절대값에 어떤 작은 양의 숫자를 곱해도 f(x)보다 크게되는 순간이 x를 키우다보면 언젠가는 나타난다.**  
```
A. 두 함수 f(n), g(n)에 대해 g(n) 이 f(n) 에 비해 증가율이 높을 경우 f(n) ∈ o(g(n)) 이라고 표기합니다.
수학적으로 정의하면  가 될 경우 f(n) ∈ o(g(n)) 가 되는 것이죠.
이것은 g(n)의 증가율이 f(n)의 증가율보다 높다는 것을 보여 주기 위해 가끔 사용되는 표기법입니다.

B. 예를 들면 2n ∈ o(n²) 이지만 2n² ∈ o(n²) 은 아닙니다. 아닌 이유는 A 에서의 정의상 극한값이 0이
안되기 때문이죠. 그것은 두 함수 2n² 와 n² 의 증가율이 같다는 의미입니다
```
