## 5.1 이진탐색  

1차원 리스트에 데이터가 정렬되어 있을 때 주어진 데이터를 효율적으로 찾는 알고리즘.  
데이터 정렬 되있으면 -------------------> 최악의경우 logN번만 비교 
정렬 안되있으면 순차탐색을 먼저 수행한다.---> 최악의 경우 O(N)시간  
(N = 리스트에 저장된 항목수)  

다음은 이진탐색 알고리즘  

```
binary_search(left, right, t)  
[1] if leff > right: return None # 탐색 실패 (즉, t가 리스트에 없음)
[2] mid = (left+right) //2       # 리스트에서 탐색할 부분의 중간 항목의 인덱스 계산  
[3] if a[mid] == t: return mid   # 탐색 성공
[4] if a[mid] > t: binary_search(left,mid-1,t)  # 앞부분 탐색 
[5] else: binary_search(mid+1, right t)         # 뒷부분 탐색   

```  

**수행시간**  
```
1번 비교시 N/2개원소 남음 
2번      N/(2^2)
...
k번      N/(2^K) = 1개 마지막엔 남고 찾았다면 

N = 2^K
K = log N          
       2

K가 시행 횟수이기 때문에 O(log N)
                         2


(혹은 단순하게 이렇게 나열해봐도 된다. 

1      0
2      1
3      2

...
N      log N
          2  

```  

**단점**  
데이터의 삽입과 삭제가 빈번하면 정렬을 유지가하기 시간이 걸린다.  
1회의 삽입, 삭제 연산 수행 시 최악경우 O(N) 시간이 소요된다.  

````python
# 예전에 코드로 구현한것 기억해서해봄.  

def binary_search(start,end,target,lista):
#     print(start)
#     print(end)
    
    if  start > end:
        return -1
    
    mid = (start+end)//2
#     print(mid)
    
    if lista[mid] == target:
        return mid
    
    elif lista[mid]> target:
        result = binary_search(start,mid-1,target,lista)
        
    elif lista[mid]< target:
        result = binary_search(mid+1, end,target,lista)
        
    return result
        

list1 = [5,8,15,20,25,30,40,50,54,66,69,83,86,90]
start_index = 0
end_index=len(list1)-1

value = binary_search(start_index, end_index,83,list1)
print(value)
    
````

이진탐색트리에서 탐색, 삽입, 삭제 연산은 
공통적으로 루트에서 탐색을 하기 시작하여 최악의 경우에 이파리 까지 내려가고,  
삽입과 삭제 연산은 다시 루트까지 거슬러 올라가야한다.

트리를 한층 내려갈 때는 재귀호출이 발생하고, 한 층을 올라갈 때는 부모와 자식노드를

[log (N+1)] ~~ logN <= h <= h

````python
### 5-3 avl.py

class Node:
    def __init__(self,key,value, height, left=None, right=None):
        self.key =key
        self.value = value
        self.height = height
        self.left = left
        self.right = right
    
class AVL:
    def __init__(self):
        self.root = None
    
    def height(self,n):
        if n == None:
            return 0
        
        return n.height
    
    def put(self,key, value): # 삽입연산 
        self.root = self.put_item(self.root,key,value)
    
    def put_item(self,n,key,value):
        
        if n == None:
            return Node(key,value,1)
        
        elif n.key > key:
            n.left = self.put_item(n.left,key,value)
    
        elif n.key < key:
            n.right = self.put_item(n.right,key,value)
            
        else:
            n.value = value
            return n
        
        n.height = max(self.height(n.left) , self.height(n.right)) + 1
            
        return self.balance(n)
        
        
    
    def balance(self,n): # 불균형 처리 
        
        if self.bf(n) > 1:
            if self.bf(n.left) <0:
                n.left = self.rotate_left(n.left)
            n = self.rotate_right(n)
        
        elif self.bf(n) < -1:
            if self.bf(n.right) > 0:
                n.right = self.rotate_right(n.right)
            n = self.rotate_left(n)
            
        return n 
        
        
    
    def bf(self,n): # bf 계산 
        return self.height(n.left) - self.height(n.right)
    
    def rotate_right(self,n): # 우로회전
        x = n.left
        n.left = x.right
        x.right = n
        
        # x.rigt = n 먼저 쓸시 애러뜨니 주의해라.---> n이 n을 가리키는 상황이됨.
        
        n.height = max(self.height(n.left), self.height(n.right)) + 1 
        x.height = max(self.height(x.left), self.height(x.right)) + 1 
        
        return x 
    
    def rotate_left(self,n): # 좌로회전 
        x = n.right
        n.right = x.left
        x.left = n
        
        n.height = max(self.height(n.left), self.height(n.right)) + 1 
        x.height = max(self.height(x.left), self.height(x.right)) + 1 
        
        return x
        
    
    def delete(self,key): # 삭제연산 
        pass
    
    def minn(self): #최소값 찾기 
        pass
    
    def inorder(self,n):
        if n == None:
            return
        
        self.inorder(n.left)
        print(n.key," ",end="")
        self.inorder(n.right)
        
    def preorder(self,n):
        if n == None:
            return
        
        print(n.key," ",end="")      
        self.preorder(n.left)
        self.preorder(n.right)
    
    
````
