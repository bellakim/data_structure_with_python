### 3.1 스택  
```
스택은 한쪽끝에서만 항목(item)을 삭제하거나 새로운 항목을 저장하는 자료구조이다.  
```  

push연산 - 새로운 항목 저장  
pop연산 - 항목을 삭제  

**파이썬 리스트로 표현된 스택**  
( 단순히 각 항목이 리스트 원소에 저장된것과 비슷)
<img src="images/IMG_20181106_133225.jpg" width="500">  

**단순연결리스트로 구현된 스택**  
top - 스택의 가장 위에 있는 항목을 참조하는 래퍼런스   
스택이 empty일 때 top = None 이다.  
<img src="images/IMG_20181106_133212.jpg" width="500">




````python
# 프로그램 3-1 파이썬 리스트로 구현한 스택  

def push(item): # 삽입연산 
    stack.append(item)
    
def peek(): # top 항목 접근  
    if len(stack) !=0:
        return stack[-1]
    
def pop(): # 삭제 연산  
    if len(stack) != 0:
        item = stack.pop(-1)
        return item
    
stack = []
push('apple')
push('orange')
push('cherry')
print('사과, 오랜지, 체리 push 후:\t', end="")
print(stack, '\t < - top')  
print('top 항목: ', end="")
print(peek())
push('pear')
print('배 push 후:\t', end="")
print(stack, '\t < - top')
pop()
push('grape')
print('pop(), 포도 push 후:\t', end="")
print(stack, '\t<- top')
````

````python
# 프로그램 3-2 단순연결리스트로 구현한 스택  

class Node: # Node 클래스
    def __init__(self,item,link):
        self.item = item
        self.next = link
    
def push(item): # push 연산
    global size
    global top
    
# 예전 연결 리스트처럼 해봄. 
#     this_node = Node(item,None)
    
#     if size==0:
#         top = this_node
#         size+=1
#     else:
#         this_node.next = top
#         top= this_node

#         size+=1

#     책에나온대로 풀면( 아래서 top에 미리 None할당해서 이것 가능하다.)
    top = Node(item, top)
    size +=1

    
def peek(): # peek 연산  
    if size != 0:
        return top.item
    
    
def pop(): # pop 연산 
    global size
    global top
    
    if size !=0 : 
        top_item = top.item
        top =top.next
        size-=1
        return top_item
        
def print_stack(): # 스택 출력 
    
    print('top ->\t', end="")
    p = top
    while p:
        if p.next != None:
            print(p.item, '->', end ="")
        else:
            print(p.item, end="")
        p = p.next
    print('')
    
#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@222222@@
top = None
size = 0

push('apple')
push('orange')
push('cherry')
print('사과, 오랜지, 체리 push 후:\t', end="")
print_stack()
print('top 항목: ', end="")
print(peek())
push('pear')
print('배 push 후:\t', end="")
print_stack()
pop()
push('grape')
print('pop(), 포도 push 후:\t', end="")
print_stack()



````

#### 수행시간 

파이썬의 리스트로 구션한 스택 --->push. pop 연산 각각 O(1)시간 소요되지만, 파이썬 리스트는 크기가 동적으로 확대, 축소되며 이때 모든 항목들을 새로운 크기가진 리스트로 복사되기 때문에 O(N)시간이 소요된다.   

단순연결리스트로 구현한 스택 ---> push, pop 연산은 각각 O(1) 시간이 걸린다. (연결리스트의 맨 앞 부분에서 노드를 삽입하거나 삭제해서)  


### 3.2 스택의 응용  

1.괄호 짝 마추기 문제  
```
왼쪽 괄호는 스택에 push, 오른쪽괄호를 읽어며 pop수행  
입력의 모든 괄호들을 에러 없이 처리한 후에 스택이 empty ---> 입력괄호 모두 짝 맞는다.
```

2.회문 문제  
```
전반부의 문자들을 스택에 push 한 후, 후반부의 각 문자를 차례로 pop한 문자와 비교  
마지막 비교까지 두문자 동일, 스택이 empty되면 문자열은 회문이다.   
(문자열 길이 홀수인경우 중간 문자만 읽고 바로 버린다.)  
```  

**응용**  
미로찾기, 트리 순회, 그래프 탐색, 함수호출 등 

### 3.3 큐  

```
큐는 삽입과 삭제가 양끝에서 각각 수행되는 자료구조이다.
```
선입선출  

```
여기에 삽입 --->   rear             front   <-----여기서 삭제(처리)
                  l                l 
                  []  []  []  []  []   
```

큐는 뒤(rear)에서 삽입하고, 앞(front)에서 삭제 연산을 수행해서, 큐의 가장 앞의 항목을 가리키는 변수인 front와 가장 뒤에 있는 항목을 가리키는 변수인 rear가 필요하다.  
큐는 선형 자료구조여서 파이썬의 리스트, 단순연결리스트로 구현 가능하다.  

````python
# 프로그램 3-3  파이썬 리스트로 구현한 큐  

def add(item): # 삽입연산 
    q.append(item)
    
    
def remove(): # 삭제연산
    if len(q) != 0:
        item = q.pop(0)
        return item
    

def print_q(): # 큐 출력  
    print('front -->', end ="")
    
    for i in range(len(q)):
        print(f'{q[i]}  ', end="")
    
    print(' <--rear')

    
q = []
add('apple')
add('orange')
add('cherry')
add('pear')
print('사과, 오렌지, 체리, 배 삽입 후: \t', end='')
print_q()

remove()
print('remove한 후: \t', end="")
print_q()  
remove()
print('remove한 후: \t', end="")
print_q()  

add('grape')
print('포도 삽입 한 후: \t', end="")
print_q()  
````

````python
# 프로그램 3-4 단순연결리스트로 구현한 큐 

class Node:
    def __init__(self,item,link):
        self.item = item
        self.next = link
        
def add(item): # 삽입연산
    global rear
    global front
    global size 
    
#     # 내가한것 
#     added_node = Node(item,None)
#     if size >1 :
#         rear.next = added_node
#         rear = added_node
    
#     elif size ==1:
#         rear = added_node
#         front.next = rear
#     else:
#         front = added_node
#     size +=1


    # 책에 나온대로 
    new_node = Node(item, None)
    if size == 0:
        front = new_node
    else:
        rear.next = new_node
    
    rear = new_node # 책에서는 이렇게 최초노드에 rear, front 둘다 줬다.
    size+=1
    
    
def remove(): # 삭제연산  

    global rear
    global front
    global size     
    
    if size !=0:
        fitem = front.item
        front = front.next
        size -= 1
        if size == 0:
            rear == None
        
        return fitem

def print_q(): # 큐 출력
    
    p = front

    print('front:\t', end="")
    
    while p:
        
        if p.next !=None:
            print(p.item, '-> ', end = "")
        else:
            print(p.item, end ="")
        p = p.next
    print(' :rear')

# 초기화 3개
size = 0
rear = None
front =None

add('apple')
add('orange')
add('cherry')
add('pear')
print('사과, 오렌지, 체리, 배 삽입 후: \t', end='')
print_q()

remove()
print('remove한 후: \t', end="")
print_q()  
remove()
print('remove한 후: \t', end="")
print_q()  

add('grape')
print('포도 삽입 한 후: \t', end="")
print_q()  



````

#### 응용 

```
큐 자료구조는 CPU의 태스크 스캐줄링, 네트워크 프린터, 실시간 시스템의 인터럽트, 이벤트 구종방식, 컴퓨터 시뮬레이션, 콜센터 전화 서비스 처리등에 사용  
이진트리의 레벨순회 와 그래프의 너비 우선탐색에 이용된다.
```

#### 수행시간  

리스트로 구현한 큐의 add, remove :  O(1) 시간이 소요된다. 하지만 리스트 크기를 확대, 축소 시키는 경우 큐의 모든 항목들을 새 리스트로 복사해야 하므로 O(N) 시간이 소요된다.  
단순연결리스트로 구현한 큐의 add, remove : O(1) 시간이 걸린다.(삽입, 삭제 연산이 rear, front로 인해 연결리스트의 다른 노드 일일이 방문 할 필요 없이 연산 수행되어서.)  

### 3.4 데크  


````python
# 프로그램 3-5 데크 프로그램  

# collection 페키지에 데크 정의되 잇다.  
# 삽입,삭제 등의 연산은 리스트의 연산들과 유사.

from collections import deque  
dq = deque('data')
for elem in dq:
    print(elem.upper(), end="")
print() 

dq.append('r')
dq.appendleft('k')
print(dq)

dq.pop()
dq.popleft()
print(dq)

print(dq[-1])
print('x' in dq)
dq.extend('structure')
dq.extendleft(reversed('python'))
print(dq)
````
