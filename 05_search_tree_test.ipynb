이진탐색트리? 
이진탐색트리는 각 노드에 자식이 2개이며 왼쪽자식의 키가 각 노드의 키보다 작고 오른쪽 자식의 키가 각노드의 자식보다 크다. 또한 키값의 중복을 허용안하고 균형 이진 탐색트리 는 더 세부이기 때문에 이진탐색트리는 편향될 수도 있다.  
그래서 높이가 logN(완전이진탐색트리의 경우) <= H < N(편향이진 탐색트리인 경우) 이다. 

균형이진트리인데 위같은 속성과 같은 특징 (자식키 -부모키 관계) 가지고 있기때문에 이트리를 이진탐색트리로 봐도 될 것 같다.

n.right = None인 노드를 찾으면 되는데 이는 높이만큼만(OlogN) 내려가면 된다.  
좀더 정확히 설명하면?  

균형 이진탐색트리의 경우 A(H)를 높이 H인 균형이진탐색트리를 만들기 위한 최소 노드의 수라고 할때 A(H) = A(H-1) + A(H-2) + 1 이기 때문에  이는 피보나치 ㅎ함수를 닮아서 
A(H) >= F(H) =  e^H/root(5) 이다.

높이 H인 균형이진탐색트리를 만들기위한 최소 노드의 수가 N이다를 다시 말하면  
노드 N개로 만들 수 있는 균형이진탐색트리의 최대 높이가 H 이다.  

즉 A(H) = N 이라 할때 
```
N >= e^H/root(5)  --> 1.44log N  > = H 이므로 O(logN) > = H 이다.  
                             2 
```

AVL?  
균형이진탐색트리.  
이진트리를 탐색할때 이진트리에 계속 점진적 큰값들이 삽입되서 편향되게 트리가 구성될 경우 높이가 N이 되서 탐색시 이진탐색트리로서의 이점이 사라진다.  그래서 만든것이 균형이진 탐색트리인데 이 트리의 조건은 개별 노드에서 좌우 서브트리의 높이 차이가 2가 안되는 트리이다.  AVL트리에서는 이를 위해서 어떤 삽입, 삭제 등 트리에 변동이 있고난후 균열이 일어난 (좌우 서브트리 높이가 다르다고 판단되는 노드)의 좌측 혹은 우측을 기존의 균열일어난 노드로 올려버리고 그 좌,혹 우 서브트리를 부모트리의 좌,혹우 에 연결해 버리는 회전연산을 통해 기존의 2이상 차이나던 서브트리의 높이차를 줄인다.  
최대 높이의 경우 5.13에 설명한 바와같이 logN정도여서.. 줄어든 탐색시간을 가능하게 한다.

````python
# 5.25  

count = 0

class Node:
    def __init__(self,key,value, left=None, right=None):
        self.key = key
        self.value = value
        self.left = left
        self.right = right
        
class BST:
    def __init__(self): # 트리 생성자
        self.root = None
    
    
    def get(self,k): # 탐색 연산  
        result = self.get_item(self.root,k)
    
        return result
    
    def get_item(self,n,k):
        
        if n== None:
            return 
            
        if n.key > k:
            value = self.get_item(n.left,k)
    
        if n.key < k:
            value = self.get_item(n.right,k)
            
            
        if n.key == k:
            return n.value
            
        return value
        
    
    def put(self, key, value): # 삽입 연산)
        self.root = self.put_item(self.root,key,value)
        

    def put_item(self,n,key, value):
        
        if n == None:
            n = Node(key,value,left=None,right=None)
        
        elif n.key > key:
            n.left = self.put_item(n.left,key,value)
            
        elif n.key < key:
            n.right = self.put_item(n.right,key,value)
        
        elif n.key == key:
            n.value = value
            
        return n 
    
# k 번째 작은 키를 찾는
# 즉 중위순회하면서 count해서 몇번째 값인지 찾아오자.  

# 전역변수 사용 
    def kth_smallest1(self,k):
        global count 
        
        if self.root == None:
            print('비어있다.')
            return 0
        
        result = self.kth_count1(self.root,k)
#         print(count)
    
    def kth_count1(self,n,k):
        
        global count
        global result
        
        if n == None:
            return 
    
        self.kth_count1(n.left,k)
#         print(n.key," ",end="")
        if k == count:
            print(f'{count}, {n.key}')
            result = (n.value, n.key)
        count = count + 1 
        self.kth_count1(n.right,k)
        
# 전역변수 사용 안하고 값 전달해 가면서 사용  

    def kth_smallest2(self,k):
        
        if self.root == None:
            print('비어있음')
            return 0
        
        result = self.kth_count2(self.root,k,0,False,None)
        
        return result
    
    
    def kth_count2(self,n,k,count,found,output):
        if n == None:
            return (count,found,output)
        
        # 찾았으면 이 아래는 실행 안한다 
        # ~count 자연스래 안됨.
        if found == True:
            return (count,found,output)
        
        count,found,output = self.kth_count2(n.left, k, count, found,output)

        
        
        if count == k and found==False:
            found = True   
            output = n.key
        
        if count != k:
            count = count +1
            
        
        
        count,found,output = self.kth_count2(n.right,k,count,found,output)

        return (count,found,output)
    
            

        
        

# 책의 내용 따라가 보면 
    def minn(self): # 최솟값 가진 노드 찾기.

        if self.root == None:
            return None
        
        return self.minimum(self.root)
    
    def minimum(self,n):
        if n.left == None:
            return n
        return self.minimum(n.left)
        

# 책보고 다시 해봄. 
    def delete_min(self): # 최솟값 삭제  
        if self.root == None:
            return None
        
        self.root = self.del_min(self.root)
        
    
    def del_min(self, n):
        if n.left == None:
            return n.right
        else:
            n.left = self.del_min(n.left)
        
        return n


        
# 책보고 다시 해봄.
    def delete(self,k): # 삭제 연산 
        self.root = self.del_node(self.root,k)
    
    def del_node(self,n,k):

        if n == None:
            return None

        if n.key == k:
        
            if n.left == None:
                return n.right
            if n.right == None:
                return n.left
            
            target = n
            n = self.minimum(target.right) #중위순회 다음타자.
            n.right =  self.del_min(target.right)
            n.left = target.left            

        elif n.key > k:
            n.left = self.del_node(n.left,k)
            
        elif n.key < k:
            n.right = self.del_node(n.right,k)
        
        
        return n
        
        
    # 전위순회 
    def pre_order(self,n):
        
        if n == None:
            return
        
        print(n.key," ", end="")
        self.pre_order(n.left)
        self.pre_order(n.right)
        
        
    # 중위순회 
    def inorder(self,n):
        
        if n == None:
            return
        self.inorder(n.left)
        print(n.key," ",end="")
        self.inorder(n.right)
        
        
    
    
    
    
````

````python
# 5.28 AVL 트리의 delete() 메소드를 작성하시오.  

### 5-3 avl.py

class Node:
    def __init__(self,key,value, height, left=None, right=None):
        self.key =key
        self.value = value
        self.height = height
        self.left = left
        self.right = right
    
class AVL:
    def __init__(self):
        self.root = None
    
    def height(self,n):
        if n == None:
            return 0
        
        return n.height
    
    def put(self,key, value): # 삽입연산 
        self.root = self.put_item(self.root,key,value)
    
    def put_item(self,n,key,value):
        
        if n == None:
            return Node(key,value,1)
        
        if n.key > key:
            n.left = self.put_item(n.left,key,value)
        if n.key < key:
            n.right = self.put_item(n.right,key,value)
            
        else:
            n.value = value
            return n
        
        n.height = max(self.height(n.left) , self.height(n.right)) + 1
            
        return self.balance(n)
        
        
    
    def balance(self,n): # 불균형 처리 
        
        if self.bf(n) > 1:
            if self.bf(n.left) <0:
                n.left = self.rotate_left(n.left)
            n = self.rotate_right(n)
        
        elif self.bf(n) < -1:
            if self.bf(n.right) > 0:
                n.right = self.rotate_right(n.right)
            n = self.rotate_left(n)
            
        return n 
        
        
    
    def bf(self,n): # bf 계산 
        return self.height(n.left) - self.height(n.right)
    
    def rotate_right(self,n): # 우로회전
        x = n.left
        n.left = x.right
        x.right = n
        
        # x.rigt = n 먼저 쓸시 애러뜨니 주의해라.---> n이 n을 가리키는 상황이됨.
        
        n.height = max(self.height(n.left), self.height(n.right)) + 1 
        x.height = max(self.height(x.left), self.height(x.right)) + 1 
        
        return x 
    
    def rotate_left(self,n): # 좌로회전 
        x = n.right
        n.right = x.left
        x.left = n
        
        n.height = max(self.height(n.left), self.height(n.right)) + 1 
        x.height = max(self.height(x.left), self.height(x.right)) + 1 
        
        return x
        
  
    # 삭제연산 
    # 1단계: 이진탐색트리에서와 동일한 삭제 연산을 수행  
    # 2단계: 삭제된 노드로부터 루트 방향으로 거슬러 올라가며
    #       불균형이 발생한경우 적절히 회전 연산을 수행.  
    
    def delete(self,key): # 삭제연산 
        if self.root == None:
            return None
        
        self.del_node(self.root,key)
        
    def del_node(self,n,key):
        
        if n == None:
            return None
        
        
        if n.key > key:
            n.left = self.del_node(n.left,key)            
        if n.key < key:
            n.right = self.del_node(n.right,key)
        
        # 키값 찾으면 삭제 
        if n.key == key:
            # 자식 1개 혹은 None일때 
            if n.left == None:
                return n.right
            if n.right == None:
                return n.left
                
            # 자식 2개일때 위에서 retun안되고 이하로 내려감.  
            #중위후계자 
            target = n 
            n = self.mininmum(target.right)
            n.right = self.del_node(target.right)
            n.left = target.left  
            
        # 높이 최신화 
        n.height = max(self.height(n.left),self.height(n.right))
        return self.balance(n)
        
        # 여기서 깨닫게 된것. 
        # 여기서 balance에 대입하는 n이라는것은 
        # 키값을 찾은 그곳이 아니다. 
        # 키값 찾은뒤 해당 node부터 root까지 타고 올라가는 경로에 있는 모든 root이다.
        # 거기서 언제 bf가 2 이상되서 balance에서 회전연산 일어 나는것은
        # 내가 초기에 root부터 2이상 높이차를 좌우 찾아가는게 아니라 
        # 끝부터 2이상 좌, 우 자식 서브트리 높이 차 최초로 나는 지점 찾고 
        # 그 바로 아래 높이차 경향 따라 rr,rl등을 결정해 주는것이었다. 
        
        
        
        
    
    def minn(self): #최소값 찾기 
        if self.root ==None:
            return None
        
        return self.mininmum(self.root)
        
    def mininmum(self,n):
        if n.left == None:
            return n
        
        # None 아닐시에 아래 왼쪽자식으로 내려간다.
        return self.mininmum(self,n.left)
    
    def inorder(self,n):
        if n == None:
            return
        
        self.inorder(n.left)
        print(n.key," ",end="")
        self.inorder(n.right)
        
    def preorder(self,n):
        if n == None:
            return
        
        print(n.key," ",end="")      
        self.preorder(n.left)
        self.preorder(n.right)
    
    
````

Splay 트리에 대해 조사하고 AVL트리, LLRB트리와 비교했을 때 어떤 장단점을 갖는지 설명하시오.

```
Splay트리

**Splay 연산은 임의의 노드 x를 루트로 만드는 연산**  

1. x가 루트이면, 루트를 만드는 데 성공했으므로 종료한다.
2. x의 부모 p가 루트이면, Rotate(x)를 행하고 종료한다. (Zig Step)
3. x의 조부모를 g라고 하면, 다음 두 가지 경우가 있다.
3-1. g→p의 방향과 p→x의 방향이 같은 경우, Rotate(p) 이후 Rotate(x)를 행한다. (Zig-Zig Step)
3-2. g→p의 방향과 p→x의 방향이 다른 경우, Rotate(x)를 두 번 행한다. (Zig-Zag Step)
4. 1로 돌아가서 루트가 될 때까지 반복한다.






삽입, 삭제, 검색 등의 쿼리를 amortized O(log n)에 처리 가능하며 Splay 연산을 이용해서 구간에 대한 쿼리가 자유롭고 AVL Tree나 Red-Black Tree와 같은 다른 Binary Search Tree보다 구현이 단순
```
