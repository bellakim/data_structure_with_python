def selection_sort(a):
    for i in range(0,len(a)-1):
        minimum = i
        
        for j in range(i,len(a)): # sub list 탐색이 i부터 시작하는것 주의

                minimum = j
        a[i],a[minimum] = a[minimum], a[i]   # 최소 인덱스만 가지고 있다가~마지막에 for문 마지막에 교환

````python
# 11-1 dpqsort.py  
# 이중피벗 퀵정렬  

def dpqsort(a,low,high):
    if high > low:  
        p1 = a[low]           # 작은 피벗 
        p2 = a[high]          # 큰 피벗  
        if p1 > p2:  
            a[low],a[high] = a[high],a[low] # 작은피벗 < 큰피벗 보장 
            p1=a[low]
            p2=a[high]
        
        i = low+1     # 3) 움직여 가면서 큰그룹~작은그룹 끼리 교환해줄 인덱스   
                      # 이것과 lt 따로 있는 이유는?  
                      # -> 그 안에 p1보다 크고 p2보다 작은 값 들어가도록. 
                      #          ( 그 사이값이면 i 가 전진해 버리는데 그러면 더이상 gt,lt와 교환에 참가 안한다.)  
                      #          ( p1보다 작은 값에 대해서 둘다 전진하는 이유는 어쨌든 그 값보다)    
                      #          ( 중간에 위치한 값이 더 크니까 둘다 더해서 전진하는것.)
                      # 중간 값인 경우 i 전진 시키고 
                      #                     그다음번 a[i]>p2이면 그곳과 a[gt]교환
                      #                            a[i]<p1이먄 a[lt]와 교환  
        lt = low+1    # 1) 작은피벗 최종위치 (이것보다 작은인덱스에 작은갑 들어가도록   
        gt = high-1   # 2) 큰 피벗 최종 위치 (이것보다 큰 인덱스에 큰값 들어가도록 
        
        while i <=gt:  
            if a[i] < p1:     # case 1  
                a[i],a[lt] =  a[lt],a[i]
                i+=1
                lt+=1   
            elif p2 < a[i]:   # case 2  
                a[i],a[gt]=a[gt],a[i]
                gt-=1  
            else:
                i +=1  
                
        lt-=1  
        a[low],a[lt]=a[lt],a[low]
        gt+=1
        a[high],a[gt]=a[gt],a[high]
        dpqsort(a,low,lt-1)
        dpqsort(a,lt+1,gt-1)
        dpqsort(a,gt,high)
            
    
a = [54,88,77,26,93,17,49,10,17,77,11,31,22,44,17,20]
print('정렬 전:\t', end="")
print(a)
dpqsort(a,0,len(a)-1)
print('정렬 후:\t', end="")  
print(a)   
````

**버블정렬의 개념을 이용하여 주어진 리스트가 정렬되어 있는지 검사하는 파이썬 함수를 작성하시오** 함수의 매개변수 리스트 a, 검사하고자 하는 부분의 가장 왼쪽 원소의 인덱스 left, 가장 오른쪽 원소의 인덱스 roght이다.  

````python
# 7.18  
# 버블 정렬 개념이용 -> 주어진 리스트 정렬되있는지 검사하는 파이썬 함수 작성.  

# 버블소트 : 
# 최선 최악의 경우 모두 n(n-1)/2 비교. 최선에서는 교환x  


import time 

# 이하는 내가 짜본 버블 소트   

def bubble_sort(list1):
    
    for i in range(len(list1)):
        for x in range(len(list1)-i-1):
            # 0~ len(list1)-i까지만 비교하겠다. 
            pre = list1[x]
            later = list1[x+1]  
            if pre > later:
                list1[x], list1[x+1] = list1[x+1],list1[x]
                
    return list1
                


# 나는 이렇게 해보겠다. 
# 교환 한번이라도 하는 순간 for문 나와서 False출력   

def is_sorted_test_with_buble_sort_concept(list1):
    
    
    for i in range(len(list1)):
        for x in range(len(list1)-i-1):
            # 0~ len(list1)-i까지만 비교하겠다. 
            pre = list1[x]
            later = list1[x+1]  
            if pre > later:
                return False
                
    return True
                
            

        
    
     
lista = [1,2,3,4,5]
listb = [1,4,3,5,6] 

import random
listc = list()
for i in range(10000):
    listc.append(random.randrange(1,1000))



print(is_sorted_test_with_buble_sort_concept(lista))
print(is_sorted_test_with_buble_sort_concept(listb))




# 임의로 만든 랜덤 10000개 원소 가진 리스트 가지고 테스트 해보겠다. 

start_time0 = time.time()  

print('위 방법 사용해서 정렬됬는지만 검사시 시간 측정 \n')
start_time1 = time.time()  
is_sorted_test_with_buble_sort_concept(listc)
print(time.time()-start_time1,'\n')


print('버블 소트시 시간 측정 \n')
start_time2 = time.time()  
bubble_sort(listc)
print(time.time()-start_time2,'\n')

print('일반 소트 시 시간 측정. \n')
start_time3 = time.time()  
listc.sort()
print(time.time()-start_time3)

# 결과 보면 실제 소트 보다 시간 더 적게 걸린다. 
````

반복 합병정렬을 위한 파이썬 프로그램을 작성하시오.  
반복 합병 정렬 :
