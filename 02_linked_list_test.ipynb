```python
# 프로그램 2-1 slist.py

class SList:
    class Node:
        def __init__(self,item, link): # 노드 생성자
            self.item = item           # 항목과 다음 노드 레퍼런스 
            self.next = link
                
    def __init__(self):                # 단순연결리스트 생성자 
        self.head = None               # head와 항목 수(size)로 구성 
        self.size = 0
        
    def size(self):
        return self.size
    
    def is_empty(self):
        return self.size == 0
        
    def insert_front(self,item):
        if self.is_empty():
            self.head = self.Node(item,None) #head가 새 노드 참조 
        
        else:
            self.head = self.Node(item,self.head)  #head가 새 노드 참조 
        self.size += 1
        # 여기서 원래 head뒤로는 똑같아서 변경안해줘도 된다. 
    
    def insert_after(self,item,p):
        p.next = self.Node(item,p.next)  ######? Slit.Node? 
        self.size += 1
            
    def delete_front(self):
        if self.is_empty():                     # empty인 경우 에러 처리 
            raise EmptyError('Underflow')
        else:  
            self.head = self.head.next   # head가 둘째 노드를 참조 
            self.size -=1
            
    def delete_after(self,p):
        if self.is_empty():
            raise EmptyError('Underflow')
        t = p.next
        p.next = t.next                  # p 다음 노드를 건너뛰어 연결 
        self.size -= 1 # 만약 p가 None이면 next도 None되는게 맞나 ? 
    
    def search(self,target):
        p = self.head                    # head로 부터 순차 탐색 
        
        for k in range(self.size):
            if target == p.item:
                return k                 # 탐색 성공시
            p = p.next
        return None                      # 탐색 실패시 
        
        
    def print_list(self):
        p = self.head
        
        while p:                         #리스트 맨 끝에선 None값이라 루프탈출 
            if p.next!=None:
                print(p.item, "->",end="")
            else:
                print(p.item)
            p = p.next                  # 노드들을 순차탐색.
        
class EmptyError(Exception):            # underflow시 에러 처리 
    pass
    
```

```python
# 프로그램 2-3 dlist.py
class DList:
    class Node:
        def __init__(self, item, prev, link):
            self.item = item
            self.prev = prev
            self.next = link
        
    def __init__(self):
        self.head = self.Node(None,None,None)
        self.tail = self.Node(None, self.head,None)
        self.head.next = self.tail
        self.size = 0
        
    def size(self): 
        return self.size
    
    def is_empty(self):
        return self.size ==0
    
    def insert_before(self, p, item):
        t = p.prev
        n = self.Node(item, t, p)
        p.prev = n 
        t.next = n
        self.size +=1
    
    def insert_after(self,p,item):
        t = p.next
        n = self.Node(item,p,t)
        t.prev = n
        p.next = n
        self.size += 1
    
    def delete(self,x):
        f = x.prev
        r = x.next
        f.next = r
        r.prev = f
        self.size -= 1
        return x.item
    
    def print_list(self):
        if self.is_empty():
            print('리스트 비어있음')
        else:
            p = self.head.next
            while p != self.tail:
                if p.next != self.tail:
                    print(p.item, '<=>', end="")
                else:
                    print(p.item)
                p = p.next
        

class EmptyError(Exception):
    pass
    
```

```python
# 2.9  
# 주어진 단순연결리스트를 역방향 단순연결리스트로 만드는 함수를 작성하시오

# 변수들만 사용하고 리스트에 저장하는 방식 안쓰고 해보고싶다. 
# https://www.pythoncentral.io/reverse-singly-linked-list/ 참고 

s1= SList()
s1.insert_front('orange')
s1.insert_front('apple')
s1.insert_after('cherry',s1.head.next)
s1.insert_front('pear')
s1.print_list()

def reverse_singly_list(lista):
    
#     초기값. 
    prev = None
    curr = lista.head # pear
    nex = lista.head.next # apple
    
    while nex != None:
        
        curr_next_temp = nex # apple #orange
        
        curr.next = prev # None # pear
        
        
        
        prev = curr # pear # apple
        curr = curr_next_temp # apple
        nex = curr_next_temp.next # orange
    
    curr.next = prev
    lista.head = curr
        
        
    
    return lista 
    # None<- pear apple->orange ->cherry ->None
    
new_list = reverse_singly_list(s1)
new_list.print_list()
```

```python
# 2.11
# 정렬된 2개의 단순연결리스트가 가지고 있는 공통된 숫자의 합을 계산하는 함수를 작성하시오
# 단, 각 단순연결리스트에는 중복된 숫자는 없다.  

s1= SList()
s1.insert_front(1)
s1.insert_front(2)
s1.insert_after(3,s1.head.next)
s1.insert_front(4)
s1.print_list()


s2= SList()
s2.insert_front(2)
s2.insert_front(3)
s2.insert_after(6,s2.head.next)
s2.insert_front(7)
s2.print_list()

def summation_common_num(list1,list2):
    
    # value와 list 한개 씩 받아서 그값이 list안에 있는지 체크하는 함수
    def is_in_this_list(value, target_list):
        
        result = False
        p = target_list.head
        
        while p.next != None:
            
            if p.item == value:
                result = True
            
            p = p.next
            
        if p.item == value:
            result = True
        
        return result
    
    
    # 둘중 하나의 리스트를 기준으로 탐색을 하되 
    # 각 인덱스에서 다른하나의 리스트를 전부 순회하며 searh해서 
    # 일치하는 값이 있는지 check하는 함수를 돌려야 한다.
    
    result_sum = 0
    
    p = list1.head
    
    while p.next != None:
        
        if is_in_this_list(p.item,list2):
            result_sum += p.item
        
        p = p.next
        
    if is_in_this_list(p.item,list2):
        result_sum += p.item
    
    
    
    return result_sum


out_put  = summation_common_num(s1,s2)
print(out_put)
```
