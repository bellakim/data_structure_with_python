```
구조

호출  
함수가 호출되면 호출한 함수의 수행에 필요한 지역 변수, 매개변수, 수행후 복귀할 주소들의 정보를 스택 프레임에 저장하며 시스템 스택에 push 하게 된다.  
시스템 스택의 최상위 top은 항상 현재 실행 중인 함수이며 스택 프레임이다.  

복귀  
함수의 실행이 끝나면 시스템 스택의 최상위 스택 프레임을 pop하여 삭제하면서 프레임에 저장된 주소를 확인하고 종료하게 된다.  
함수는 호출과 종료에 따라 반복되며, 수행 종료시 시스탬 스택은 공백 스택이 된다.   

실행순서  
프로그램이 실행을 시작하면서 메인 함수가 실행된다.  동시에 메인함수와 관련된 정보를 스택 프레임에 저장하며 시스템 스택에 삽입한다.  
그리고 메인 함수가 실행되면서 별도의 구현된 함수들이 실행된다. 함수를 담게된다면 호출과 복귀 정보를 스택 프레임에 담아 시스템 스택에 담게된다.  
이럴때는 메인 함수로 복귀하기 위한 주소 값이 스택 프레임에 저장되게 된다.  
불러들인 별도의 함수가 실행되며, 복귀 할 때는 pop하여 스택 프레임에 있는 정보를 알게 된다.  
마지막에는 메인 함수가 남게 되지만 메인 함수는 복귀할 주소가 없으므로 프로그램이 종료되며 시스템 스택은 공백 스택이 된다.  
```

### 2.2 참조에 의한 전달.

}

````python
# 3.24  
# 스택을 이용하여 주어진 스택의 항목들을 역순으로 만드는 파이썬 함수를 작성하시오 
# 단 주어진 스택은 단순연결리스트로 구현되어 있다고 가정하시오   

class Stack:
    def __init__(self):
        self.size = 0 
        self.top = None

    class Node:
        def __init__(self,item,link):
            self.item = item 
            self.next = link 

    def push(self,item):
        self.top = self.Node(item, self.top)
        self.size+=1

    def pop(self):
        if self.size>0:
            top_item = self.top.item
            self.top = self.top.next
            self.size -= 1        
            return top_item

    def peek(self):
        if self.size>0:
            return self.top.item

    def print_stack(self):
        p = self.top
        print('top --> \t', end="")
        while p:    
            if p.next != None:
                print(f'{p.item} --> \t ',end='')
            else:
                print(p.item)
            p = p.next


def reverse_stack(input_stack):
    new_stack = Stack()
    
    temp_list = []
    element = input_stack.top
    
    while element !=None:
        
        temp_list.append(element.item)
        element = element.next
        
    for item in temp_list:
        new_stack.push(item)
        
    return new_stack
    


##################################
s1  = Stack()
s1.push('a')
s1.push('b')
s1.push('c')
s1.push('d')
s1.push('e')
s1.push('f')
s1.print_stack()

s2 = reverse_stack(s1)
s2.print_stack()

````
