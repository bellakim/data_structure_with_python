**동적해싱방식**: 대용량이라면 동적해싱방법을 쓰는적절하다. 
동적해싱 방법은 재해시 과정에서 너무많은 시간 비용 들기때문에 재해시 수행하지 않고 테이블 크기를 조정하는 방식. 확장해싱, 선형 해싱이 있다.  
확장해싱은 디렉터리, 버킷을 사용해서 키값을 저장하는데 디렉터리는 메모리 상에 저장되고 버킷은 디스크상에 저장된다. 디렉터리에는 전역깊이(깊이에따라 디렉터리 크기 결정된다. 깊이많큼 비트수를  사용해서) , 입력되는 키값의 이진값에 따라 분류할때 쓰일 기준되는 인덱스 값들이 저장되어있고 버킷에는 실제 분류되 저장되는 키값이 앞서 디렉토리에 명시된 키값에 따라서 들어온다. 버킷이 에는 한도가 있다. bucket capacity 있어서 이것을 넘으면 디렉터리와, 버킷을 확장하거나 버킷만 확장한다. 둘다 확장하는 경우에는 충돌된 버킷의 깊이가 이미 전역깊이(디렉토리쪽깊이)와 같은경우 인데 기존에 있던 디렉토리 크기의 두배를 하고 기존에 분류하려 저장했던 인덱스 값들에 0과 1을 붙인 분류기준 값들을 넣어놓는다.  버킷은 
디렉터리이상 차수를 늘리지 않고 충돌전에 버킷안에 있던 키값들, 충돌일으킨 키값을 새로운 디렉토리 안의 분류기준에 따라서 분류해준다. 기존에 깊이가 디렉터리의 깊이보다 작던 다른 버킷들은 확장되지는 않지만 기존 디렉터리로부터오던 화살표(분료방향가리키던)이 2배가 된다.  버킷만 확장하는 경우는 충돌된 버킷의 깊이가 아직 디렉터리의깊이보다 작은 경우인데 이경우엔 버킷의 깊이를 증가시키고 하나더 늘린뒤 키값들을 분배해준다.  
선형해싱의 경우에는 디렉터리자체가 없고 버킷만 있는 형태인데 필요할때 그때그때 버킷을 한개씩만 늘려간다. 새롭게 추가된 버킷은 현재 키가 삽입된 버킷과 상관없이 가장 아래 순차적으로 추가된다. 키가 삽입되어야할 버킷에 공간없으면 체인으로 단순연결리스트 연결해서 충돌된 키들을 임시로 넣어놓고 추후 관련 키값을 담는 버킷 추가시 이동시킨다.  
각 버킷마다 모드가 존재하는데 n이 해당 버킷 가리키고 있는 상황에서 버킷을 분화할 때가 되면 이 모드를 늘리고 새로운 버킷은 늘어난 모드 2배된 값을 가지고 가장아래로 들어간다.

**2차 군집화**  
일단 이것이 생기는 배경에대해 말해보겠다.  
AVL , 래드블랙 트리의 경우 O(logN)의 탐색 삽입, 삭제 시간이 걸린다.  
이것보다 더 빠른것이 있을까? 해서 살펴본것이 해시테이블.  
해시테이블은 자료의 인덱스들이 table형태로 저장된 자료구조인데   
트리 구조에 비해 높이가 낮아서(높이가1이니까 ;;) 빠른 자료 탐색,삽입,삭제가 가능하다.O(1)시간. 
키값이 들어오면 해시함수를 통한 해싱을 통해 해시값을 저장할 해시 테이블의 인덱스로 반환한다.( 이런 과정을 거치는 이유는?  B,B+트리처럼 인덱싱을 위해 트리 상부에 키값을 놓거나 하면 높이가 높아지고 높이를 낮추더라도 미리 자료의range가 1~999인 경우에 대비해서 
0<x<2, 2< x<4 이런식으로 노드안에 키값들을 놓으면 미리 메모리 공간 할당한 상태여서 메모리 공간 낭비도심하고, 추후 탐색시에 None인 부분들을 거쳐서 key값이 일치하는 곳까지 도달하는데도 시간낭비도 심하다. 그렇기때문에 입력되는 전체 키 숫자에 대해  탐색,삭제,삽입이 용이할만큼<공간,시간적으로> 적절히 테이블 크기를 미리 할당해놓고 테이블 인덱스 안에 다 들어가도록 해시 값들은 (주로 테이블의 크기로 간편히 나눠서 )변환하는 것이다.  
하지만 이때 문제가 있는데  
키값이만약 1,11이고 테이블 크기가 10인경우 10으로 나눈 나머지가 둘다 1 이므로 같은 테이블 원소에 해당 키값 을 둘다 저장하려고 한다.  
이경우 충돌이 일어났다고 표현을 하고 충돌 해결 방법은 크게 개방주소방식, 폐쇄주소방식이 있다.  개방주소방식은 헤시테이블내의 공간을 열린공간으로 본다. 즉 해당 인덱스에서 충돌이 일어나면 기존에 있던 주소 외에도 특정 처리를 해서 다른 원소중 비어있는 원소에 넘치는 키값을 저장한다. 폐쇄주소 방식의 경우는 해당 인덱스에대해 충돌이 일어나도 그 인덱스에 저장을 한다.(체이닝, 2방향체이닝)  

여기서 비교를 하게될 2차조사방식, 랜덤조사방식은 개방주소방식 중 하나이고 그다음 원소를 저장하기위한 인덱스를 찾는 과정에 차이가 있다.  
2차 조사 방식은 어떤 인덱스에서 충돌이 일어났을때 다음 인덱스를 찾기위해 점프할때 점프하는 인덱스 값이 제곱으로 증가한다.  이경우 선형 조사방식때 야기되는 1차군집화와 비슷한 2차군집화가 발생하는데 점프하는 값에 어떤 규칙이 있으면 어떻게든 군집화가 발생하는것 같다. 왜냐하면 만약 인덱스 1에서 충돌이 발생해서 1+1^2=2 에 값을 저장하고 그다음 또 충돌나서 1 + 2^2= 5 에 값을 저장을 했다 했을때 2에서 충돌난 경우에도 그에 인접한 2+1^2 = 3, 2+2^2 = 6에 값을 저장하면서 군집이 형성 되기 때문이다.

````python
# 6.17 랜덤조사를 위한 프로그램 6-5 rand.py를 완성시키시오    

import random



class RandomProbing:                
    def __init__(self,size):    #객체의 생성자   
        self.M = size           # 테이블크기 M
        self.a = [None]*size    # 헤시테이블 a
        self.d = [None]*size    # 데이터 저장용 d
        self.N = 0 
                    
    def hash(self,key):     # 나눗셈 해시함수 
        return key%self.M

    # 책에서 대로 해봄.  

    def put(self,key,data):
        initial_position = self.hash(key)
        i = initial_position
        random.seed(1000)

        while True:
            if self.a[i] == None:
                self.a[i] = key
                self.d[i] = data
                self.N +=1
                return

            if self.a[i] == key : # data만 갱신  
                self.d[i] = data
                return
            
            j = random.randint(1,99)
            i = (initial_position +j)%self.M
            
            if self.N >= self.M:
                return
        return
        
# 책보고 다시한거   
    def get(self,key): # 탐색연산  
        initial_position = self.hash(key)
        i = initial_position
        random.seed(1000) 
        
        count = 0 
        while self.a[i]!=None:
            if self.a[i] == key:
                return self.d[i]
            
            j = random.randint(1,99)
            i = (initial_position + j) % self.M
            
            count +=1
            if count >100:
                return None
            
        return None
    
    
    def print_table(self): #해시테이블 출력  
        
        print('\n')
        for i in range(self.M):
            print(f'{i}', " ", end="")
        print('\n')
        
        for i in range(self.M):
            print(f'{self.a[i]}', " ", end="")
        print('\n')
        
````

````python
# 6.18 이중해싱을 위해 파이썬 프로그램을 작성하기오  (책의 예제처럼 M=13으로 해보자.)
# 단 h(key) = key %13, d(key) = 7-(key%7)


class DoubleHash:
    def __init__(self,size):
        self.M = size
        self.a = [None]*size  # key 저장하는 리스트 
        self.d = [None]*size  # data 저장하는 리스트
        
    def h_hash(self,key):
        return key % self.M
        
    def d_hash2(self,key):
        return (self.M //2 +1) - key % (self.M //2 +1)
        
    def put(self,key,item):
        h_index = self.h_hash(key)
        j = 0
        
        count = 0
        while True:
            index = (h_index +j*self.d_hash2(key))%self.M
            
            if self.a[index] == None:
                self.a[index] = key
                self.d[index] = item
                return
            
            if self.a[index] == key:
                self.d[index] = item
            j +=1
            
            count +=1
            if count == self.M*10:
                return
            
    def get(self,key):
        h_index = self.h_hash(key)
        j = 0
        
        count = 0        
        
        while True:
            index = (h_index +j*self.d_hash2(key))%self.M
            
            if self.a[index] == key:
                return
            
            j +=1
            
            count +=1
            if count == self.M*10:
                return        

    
    
    def print_table(self): #해시테이블 출력  
        
        print('\n')
        for i in range(self.M):
            print(f'{i}', " ", end="")
        print('\n')
        
        for i in range(self.M):
            print(f'{self.a[i]}', " ", end="")
        print('\n')
            
            
````

````python
# 6.19 2방향 체이닝을 파이썬 프로그램으로 구현해 보시오
# 2방향 체이닝은 해시 계산 두가지로 해서 
# 두 index중 리스트가 짧은 곳에 저장한다. 
# 1방향 체이닝과 다르게 각 파이썬 리스트의 엘레멘트에 들어있는 연결리스트의 길이를 저장을 어떻게 할지..생각해봐 
# -> a.d말고 따로 i를 만들어서 길이만 저장하고 최신화 하겠다. 

# 일단 기본 체이닝 작성하고 그걸 2웨이로 만들어보자. 
import random

class TwoWayChaining:
    

    class Node:
        def __init__(self,key,value,link):
            self.key = key
            self.next = link
            self.data = value
        

            
    def __init__(self,size):
        self.M = size
        self.a = [None]*size
        self.h_table_elements_lists_lengths = [0]*size
        self.d_table_elements_lists_lengths = [0]*size
        
        
    def h_hash(self,key):
        return key % self.M
    
    def d_hash(self,key):
        random.seed(999)
        key2 = random.randint(1,99)
        key3 = random.randint(1,99)
        return (key2*key +key3)%self.M
    
    def put(self,key,value):
        index_h = self.h_hash(key)
        index_d = self.d_hash(key)
        
        h_length = self.h_table_elements_lists_lengths[index_h]
        d_length = self.d_table_elements_lists_lengths[index_d]
        
        if h_length > d_length:
            index = index_d
            self.d_table_elements_lists_lengths[index_d] += 1
            print(key, index,'index_d')
        else:
            index = index_h
            self.h_table_elements_lists_lengths[index_h] += 1
            print(key, index,'index_h')
        
        
        p = self.a[index]
        while p != None:
            if p.key == key:
                p.data = value
                return
            p = p.next
            
        self.a[index] = self.Node(key,value,self.a[index]) 
        
        
    
    def print_table(self):
        
        for index,element in enumerate(self.a):
            print(index, ":",end="")
            p=element
            
            while p:
                print(f'-> [ {p.key},{p.data} ]',end="")     
                p = p.next
            print("\n")
            
            
    def get(self,key):
        index_h = self.h_hash(key)
        index_d = self.d_hash(key)
        
        # 두개 리스트 다 뒤져야 한다. 
        p = self.a[index_h]
        
        while p:
            if p.key == key:
                return p.data
            p = p.next
            
        p = self.a[index_d]
        
        while p:
            if p.key == key:
                return p.data
            p = p.next
            
        return None
        
        
        

    
````

````python
t = TwoWayChaining(13)
t.put(25,'cherry')
t.put(37,'banana')
t.put(18,'lime')
t.put(55,'mango')
t.put(22,'watermelon')
t.put(35,'orange')
t.put(50,'apple')
t.put(60,'apple')

             
t.print_table()

result = t.get(35)
print(result)
````
