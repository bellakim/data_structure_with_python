# 1.스택 

````python
# 스택은 접시 쌓듯이 맨 위에 새로운거 쌓이고 뺄때도 맨위에서 빼는 자료구조 
# 후입선출 
#  
#   초반      top -> None 
#   top -> ( 여기 쌓이고, 삭제 둘다 일어남)   -> (item,next) -> 

# 단순연결리스트로 짜보겠슴. 

# 필요한 함수 
#  class Node
#  push, pop (삽삭)
#  peek, print()
# 

# 1.코딩  
class Node: # Node 클래스
    def __init__(self,item,link):
        self.item = item
        self.next = link
    
def push(item): # push 연산
    global size
    global top
    
# 예전 연결 리스트처럼 해봄. 
#     this_node = Node(item,None)
    
#     if size==0:
#         top = this_node
#         size+=1
#     else:
#         this_node.next = top
#         top= this_node

#         size+=1

#     책에나온대로 풀면( 아래서 top에 미리 None할당해서 이것 가능하다.)
    top = Node(item, top)
    size +=1

    
def peek(): # peek 연산  
    if size != 0:
        return top.item
    
    
def pop(): # pop 연산 
    global size
    global top
    
    if size !=0 : 
        top_item = top.item
        top =top.next
        size-=1
        return top_item
        
def print_stack(): # 스택 출력 
    
    print('top ->\t', end="")
    p = top
    while p:
        if p.next != None:
            print(p.item, '->', end ="")
        else:
            print(p.item, end="")
        p = p.next
    print('')
    

top = None
size = 0

push('apple')
push('orange')
push('cherry')
print('사과, 오랜지, 체리 push 후:\t', end="")
print_stack()
print('top 항목: ', end="")
print(peek())
push('pear')
print('배 push 후:\t', end="")
print_stack()
pop()
push('grape')
print('pop(), 포도 push 후:\t', end="")
print_stack()

    
# 2.시간복잡도
#   삽,삭:바로 맨위에꺼 접근해서 연결 만바꿔주므로 O(1)
#   탐색 (N)


# 3.장점: 아래 용도에서 쓸수있음.ㅋㅋ..  

# 4.단점:  새로운 입력계속 들어오고 바닥은 잔류 가능. 
# 
# 

# 5.용도: 코드에서 괄호검사시 짝 맞출때(왼쪽괄호 읽을때 push, 오른쪽 읽을때 pop) 
#                                  -> 마지막 괄호에서 empty면 짝맞는다. 
#        회문검사 -> 반 나눠서 반에선 push, 나머지에선 스텍에서 팝한것과 같은지 비교  
#                  홀수는 중간에서 버리기만함.
#        시스템 스택-> 함수 호출시 스택에 쌓여서 복귀 순서 지킬때   
#        트리순회, 그래프탐색
#        중위표기, 후위표기법 
#        뒤로가기 
````

# 2.큐 

````python
# 1.코드 

# 삭삽 양끝에서 수행 
# 선입선출 

#   여기에 삽입 ---> front                 rear <-----여기서 삭제(처리) 
#                   l                     l 
#                  [] -> [] -> [] ->[]->  [] 


#  초기   front  -> [] <- rear

# 단순연결리스트로 구현해보겠다. 


class Node:
    def __init__(self,item,link):
        self.item = item
        self.next = link
        
def add(item): # 삽입연산
    global rear
    global front
    global size 
    
    # 책에 나온대로 
    new_node = Node(item, None)
    if size == 0:
        front = new_node
    else:
        rear.next = new_node    # 왜이런지 쪼까 햇가릴텐데 
             # 잘 생각해보면   front-> [data,None]  <-rear  이상황에서  
            # 
            #     rear.next에 새값 넣고 그것이 rear가 되게 하는것.   
            #  newnode->        [new_data, None] < -rear  
            #                         ^ 
            #                         l   
            #         front -> [data,new_node]
            #           
            #            front는 그대로 처음에 들어온거 계속 가리키고있게된것. 
    rear = new_node # 책에서는 이렇게 최초노드에 rear, front 둘다 줬다.
    size+=1
    
    
def remove(): # 삭제연산  

    global rear
    global front
    global size     
    
    if size !=0:
        fitem = front.item
        front = front.next
        size -= 1
        if size == 0:
            rear == None
        
        return fitem

def print_q(): # 큐 출력
    
    p = front

    print('front:\t', end="")
    
    while p:
        
        if p.next !=None:
            print(p.item, '-> ', end = "")
        else:
            print(p.item, end ="")
        p = p.next
    print(' :rear')

# 초기화 3개
size = 0
rear = None
front =None

add('apple')
add('orange')
add('cherry')
add('pear')
print('사과, 오렌지, 체리, 배 삽입 후: \t', end='')
print_q()

remove()
print('remove한 후: \t', end="")
print_q()  
remove()
print('remove한 후: \t', end="")
print_q()  

add('grape')
print('포도 삽입 한 후: \t', end="")
print_q()  





# 2.시간복잡도
#   삽,삭:    O(1) (rear, front각각 바로 방문) 

# 5.용도:      cpu  스케줄링 
#             실시간 인터럽트 
#             이진트리 레벨순회, BFS   
# 
````
