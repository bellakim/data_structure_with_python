## 4.1 트리  

**정의**
``` 
트리는 empty 이거나 empty가 아니면 루트 R과 트리의 집합으로 거성되는데 각 트리의 루트는 R의 자식노드이다.  
단, 트리의 집합은 공집합일 수도 있다.  

```  

**용어**  

루트(Root) - 트리의 최상위에 있는 노드  
자식(Child)노드 - 노드 하위에 연결된 노드  
차수(Degree) - 자식노드 수  
부모(Parent)노드 - 노드의 상위에 연결된 노드   
이파리(Leaf) - 자식이 없는 노드 ( 단말 Termimnal, 외부 External노드 라 하며 이파리 아닌 노드를 비단발, 외부 노드라 말한다.)
형제(Sbling)노드 - 동일한 부모를 가지는 노드 
조상(Ancestor)노드 - 루트까지의 경로상에 있는 모든 노드들의 집합  
후손(Descendant)노드 - 노드 아래로 매달린 모든 노드들의 집합  
서브트리(Subtree) - 노드 자신과 후손노드로 구성된 트리  
레벨(Level) - 루트가 레벨 1에 있고, 아래층으로 내려가며 레벨이 1씩 증가한다.  레벨은 깊이와 같다.  
높이(Height) - 트리의 최대 레벨  
키(Key) - 탐색에 사용되는 노드에 저장된 정보   

```
                     루트노드 
                       A                    레벨 1 
                  /    |    \ 
높이 4           B      C      D             레벨 2 
             / | \    / \     |   
            E  F  G   H I     J             레벨 3  
           / \    |   |      / \   
          k   L   M   N     O   P           레벨 4 
```  


일반적인 트리를 메모리에 저장하려면 각 노드에 키와 자식 수 만큼의 레퍼런스를 저장해야 한다.  
트리 노드의 최대 차수(자식노드수) 가 k 라면 k개의 래퍼런스 필드를 선언해야 한다.   
```
      1  2        k 
[key][ ][ ][... ][ ]

```  
최대차수(자식노드수) 가 k인 트리에 N개의 노드가 있다면  
None 래퍼런스 수는 Nk(총 레퍼런스의 수) - (N-1) = N(k-1)+1 이다.  
따라서 k(자식노드수)가 클수록 메모리의 낭비가 심해지고 트리를 탐색하는 과정에서 None래퍼런스 확인해야되서 비효율 적이다.  

**왼쪽자식, 오른쪽형제(Left Child - Right-Sbling)** 은 이런 단점 보완해주는 자료구조이다.  
```
노드의 왼쪽 자식과 왼쪽 자식의 오른쪽 형제노드를 가리키는 2개의 래퍼런스만을 사용해서 노드를 표현한다. 

                key
         [왼쪽자식][오른쪽 형제]
                  
```

## 4.2 이진트리  

**정의**  
```
각 노드의 자식수가 2 이하인 트리.  
데이터의 구조적 관계 잘 반영 -- > 효율적 삽입, 탐색 가능하다.  
이진트리의 서브트리를 다른 이진트리의 서브트리와 교환하는것 쉽다.  
이진트리는 empty이거나 empty가 아니면 루트와 2개의 이진트리인 왼쪽서브트리와 오른쪽 서브트리로 구성된다.  
```  



**포화이진트리**(Full Binary Tree) : 모든 이파리의 깊이가 같고 각 내부노드가 2개의 자식노드를 가진 트리이다.   
**완전이진트리**(Complete Binary Tree) : 마지막 이파리를 제외한 각 레벨이 노드들로 꽉 차있고, 마지막 노드들이 왼쪽 부터 빠짐없이 채워진 트리.(포화이진트리는 완전이진트리이다)

수행시간분석을 위해 필요한 **이진트리의 속성**   
```
 * 레벨 k에 있는 최대 노드 수는 2^(k-1)이다. 단 k=1,2,3...이다.
 * 높이가 h 인 포화이진트리에 있는 노드 수는 2^h -1 이다.  
 * N개의 노드를 가진 완전이진트리의 높이는 [log (N+1)] 이다. 
                                       2           
   (높이 h인 포화이진트리에 있는 노드수 = 2^0+2^1+2^2+ ...2^(h-1) = 2^h -1) 에 로그 취해서 
```

이진트리는 파이썬 리스트 or 단순연결리스트 확장해서 각 노드에 2개의 레퍼런스를 사용하여 저장할 수 있다.  
파이썬 리스트 사용하는 경우 아래와 같이 a[0]은 사용하지 않고 트리의 레벨 1부터 차례로 내려가며 각 레벨여서는 좌에서 우로 트리의 노드들을 a[1]부터 차례로 저장한다.

다음 규칙으로 부모,자식노드가 리스트 어디에 저장된지 알 수 있다.  
```
트리에 N개의 노드가 있다고 가정하면 
*a[i]의 부모는 a[i//2]에 있다. 단, i>1이다.
*a[i]의 왼쪽 자식은 a[2i]에 있다. 단, 2i <=N이다.
*a[i]의 오른쪽자식은 a[2i+1]에 있다. 단. 2i+1 <=N이다.

```  
완전 이진트리를 저장하기 위해 위처럼 리스트 사용할 경우 따로 자식노드들을 참조할 레퍼런스를 저장할  
필요 없어서 매우 효율적이다.  
.  
.  
하지만 아래같이 편향(Skewed)이진트리를 리스트에 저장하는 경우, 트리의 높이가 커질 수록 메모리 낭비가 심하다.  
(아래 편향된 트리를 레퍼런스를 사용하여 표현한것)
<img src="images/KakaoTalk_20181113_204957591.jpg" width="600"> 



````python
# 4.1 binary_tree
class Node:
    def __init__(self,item,left=None, right=None):
        self.item = item
        self.left = left
        self.right = right
        
class BinaryTree:
    def __init__(self): # 트리 생성자. 
        self.root =None

    def preorder(self,n): # 전위순회
        if n!=None: #맨 아래 level의 Node방문시 None이라서 함수 재귀호출 안하게 된다. 
            print(str(n.item),'',end='') # 맨먼저 노드 방문 
            if n.left:
                self.preorder(n.left)
            if n.right:
                self.preorder(n.right)
              
    def inorder(self,n): # 중위순회 
        if n!=None:
            if n.left:
                self.inorder(n.left)
            print(str(n.item),'',end='')
            if n.right:
                self.inorder(n.right)
            
    def postorder(self,n): # 후위순회 
        if n!=None:
            if n.left:
                self.postorder(n.left)
            if n.right:
                self.postorder(n.right)
            print(str(n.item),'',end='')
            
        
    def levelorder(self, root): # 레벨순회 
        queue  = []         # 리스트로 큐 자료구조
        queue.append(root)
        while queue: 
            node = queue.pop(0) # 큐에서 첫 항목 삭제 
            print(str(node.item),'',end="") # 삭제된 노드 방문 
            if node.left:
                queue.append(node.left) # 왼쪽자식, 오른쪽 자식 큐에 삽입.
            if node.right:
                queue.append(node.right)
            
    
    def height(self,root): # 트리 높이 계산 
        if root == None:
            return 0
        return max(self.height(root.left),self.height(root.right)) +1
    
t = BinaryTree()
n1 = Node(100)
n2 = Node(200)
n3 = Node(300)
n4 = Node(400)
n5 = Node(500)
n6 = Node(600)
n7 = Node(700)
n8 = Node(800)
n1.left = n2
n1.right = n3
n2.left = n4
n2.right = n5
n3.left = n6
n3.right = n7
n4.left = n8 
t.root = n1

print('\n전위순회:\t', end ="")
t.preorder(t.root)
print('\n중위순회:\t', end ="")
t.inorder(t.root)
print('\n후위순회:\t', end ="")
t.postorder(t.root)

print('\n레벨순회:\t', end ="")
t.levelorder(t.root)

print('트리높이=', t.height(t.root)) 
````

## 4.3 이진트리의 연산   
1.5장 섬 관광 예제 상기해 보도록 해라.  
**전위 순회 정의정의**  
```
전위 순회는 노드 n에 도착했을 때 n 을 먼저 방문한다. 그 다음에 n의 왼쪽 자식노드로 순회를 계속한다.  
n의 왼쪽 서브트리의 모든 노드들을 방문한 후에는 n의 오른쪽 서브트리의 모든 후손 노드들을 방문한다
```
실선 화살표 따라 각노드방문 이뤄지고 점선 화살표는 노드의 서브트리에 있는 모든 노드들을 방문한 후에 부모노드로 복귀하는 것을 나타낸다.  
(복귀하는것은 프로그램에서 메소드 호출이 완료된 수에 리턴하는 것과 같은 맥락이다.)

**중위 순회 정의정의**  
```
중위순회는 노드 n에 도착하면 n의 방문을 보류하고 n의 서브트리로 순회를 진행한다. 즉, 왼쪽 서브트리의 모든 노드들을 방문한 후에 n을 방문한다. n을 방문한 후에는 n의 오른쪽 서브트리를 같은 방식으로 방문한다. 
```

**후위 순회 정의정의**  
```
후위순회는 노드n에 도착해면 n의 방문을 보류하고 n의 왼쪽 서브트리로 순회를 진행한다. n의 왼쪽 서브트리를 방문한 후에는 n의 오른쪽 서브트리를 같은 방식으로 방문한다. 그리고 마지막에 n을 방문한다. 
```

**레벨 순회 정의정의**  
```
레벨순회는 루트가 있는 최상위 레벨부터 시작하여 각 레벨마다 좌에서 우로 노드들을 방문한다.
```

**이진트리의 높이**
```
트리의 높이 = 1+max(루트의 왼쪽 서브트리의 높이, 루트의 오른쪽 서브트리의 높이)이다. 여기서 1은 루트 자신을 계산에 반영한 것이다.
```

**수행시간**  
위의 각 연산들은 트리의 노드를 1번씩만 방문하므로 O(N) 시간이 소요된다.  


**스레드 이진트리**  
위에서 상펴본 이진트리 기본연산들은 레벨순회 제외하고 모두 스택 자료구조를 사용한다.  
함수의 재귀호출은 시스템 스택을 사용하므로 스택 자료구조를 사용한 것으로 간주한다. 스택에 사용되는 메모리 공간의 크기는 트리의 높이에 비례한다.  
스택 없이 이진트리의 연산을 구현하는 방법에는두 가지가 있다.  
1. Node에 부모노드를 가리키는  레퍼런스 필드를 추가로 선언하여 순회에 사용하는 방법
2. Node의 None 레퍼런스들을 활용하는것.(즉, None레퍼런스 공간에 다음에 방문할 노드의 레퍼런스를 저장하는 것이다. 이렇게 만든 이진ㅌ리를 스레드 이진트리(Threaded Binary Tree)라고 한다.  
**정의**
```
스레드 이진트리는 각 노드의 오른쪽 None 래퍼런스를 다음 방문할 노드를 참조하게 하고, 각 노드의 왼쪽 None레퍼런스를 직전 방문한 노드를 참조하게 한  이진트리이다.
```  

<img src="images/KakaoTalk_20181113_204955312.jpg" width="600">  

N개의 노드가 있는 이진트리에는 N+1개의 None래퍼런스 필드가 있는데, 각 노드마다 2개의 레퍼런스 필드(left,right) 가 있으므로 총 2N개의 레퍼런스 필드가 존재하고 이중에서 부모-자식을 연결하는 레퍼런스는 N-1개이기 때문이다.(부모-자식 연결하는레퍼런스가 N-1개인 이유는 루트를 제외한 각 노드가 1개의 부모노드를 갖기 때문이다.)  
스레드 이진트리는 N+1개의 None래퍼런스를 활용하여 이전에 방문한 노드와 다음 방문할 노드를 가리키도록 만들어 순회 연산이 스택없이도 수행될 수 있도록 만든 리스트이다.  

스레드 이진트리는 대부분의 경우 중위순회에 기반하여 구현된다.  
스레드 이진트리는 스택을 사용하는 순회보다 빠르고 메모리 공간도 적게 차지한다는 장점을 갖지만 데이터의 삽입과 삭제가 잦은 경우 그 구현이 비교적 복잡해서 성능 안좋다.  

또한 Node 객체에 2개의 Boolean 필드를 사용하여 레퍼런스가 스레드(다음방문할 노드를 가리키는)로 사용되는 것인지 아니면 left나 right가 트리의 부모 자식 사이의 레퍼런스인지를 각각 True와 False로 표시해주어야 한다.
(위 그림은 중위순회 이진트리 예시. 점선화살표 - 직전 방문 노드를 가리키는 스레드. 실선화살표 - 다음에 방문 노드를 가리키는 스레드) 

## 4.4 이진힙  
```
이진힙(Binary Heap)은 우선순위큐 를 구현하는 가장 기본적인 자료구조.
```    

````python
# 4.3 binary_heap.py  

# 해설보고 혼자 해봄. 

class BHeap:
    
    def __init__(self,a): # 이진힙 생성자 리스트a, 항목수 N
        self.a = a
        self.N = len(a) -1
    
    def create_heap(self): # 초기 힙 만들기 
        for i in range(self.N//2,0,-1):
            self.downheap(i)
#             self.print_heap()
#             print('@@@@@@@@@@@@@@@@@@@@@@@2222')
    
    # 메소드 초기에 임의의 순서로 키가 저장되어 있는 리스트 a[1]~a[N]의 항목들을 최소힙으로 만든다.  
    # for문이 i 가 N//2부터 시작하는데, 이는 a[N//2]부터 downheap하여 최소힙을 만드는것을 의미한다. 
    
    def insert(self,key_value): # 삽입 연산
        self.a.append(key_value)
        self.unheap(self.N)
    
    def delete_min(self): # 최솟값 삭제 
        self.a[1] = self.a.pop()
        self.downheap(1)
    
    def downheap(self,i): # 힙 내려가며 힙속성 회복 

        k = i                        
        print(k)
        flag_left = 0 
        flag_right = 0
        while(True):

                
            if 2*k <= self.N:
                if a[2*k][0] < a[k][0]: # 좌가 존재할때만 현노드와 비교해준다.
                    flag_left = 1
                    
            if 2*k+1 <= self.N:  # 우가 존재할때만 현노드와 비교해준다.
                if a[2*k+1][0] < a[k][0]:
                    flag_right = 1
                    
                    
                    
                    
            # 만약 현 노드의 좌우 다 현노드 보다 작다면 둘중 승자를 가지고 현노드와 바꿔준다. 
            if flag_left == 1 and flag_right==1:
                if a[2*k][0] < a[2*k+1][0]:
                    a[k],a[2*k] = a[2*k],a[k]
                    k = 2*k
                else:
                    a[k],a[2*k+1] = a[2*k+1],a[k]
                    k = 2*k+1
                    
            # 좌만 현노드 보다 작다면 
            if flag_left == 1 and flag_right !=1:
                a[k],a[2*k] = a[2*k],a[k]
                k = 2*k
            
            # 우만 현노드 보다 작다면 
            if flag_right == 1 and flag_left !=1:
                a[k],a[2*k+1] = a[2*k+1],a[k]
                k = 2*k+1
                
            # 좌 우 다 해당하지 않는다면? 즉 위에서 self.N 보다 작은지 통과x and 현노드 위치 양호하면 
            if (flag_right == 0 and flag_left==0):
                break
                
                
            flag_left = 0
            flag_right = 0
                
            
            
            
            
            
            
    # 상향식(bottom-up)방식으로 각 노드에 대해 힙속성을 만족하도록 부모와 자식노드를 서로 바꾼다.  
    # 힙을 만들기 위해선 a[N//2]부터 a[1]까지 차례로 downheap을 각각 수행하여 힙속성을 충족시킨다.
    
    # 이때 a[N//2+1]~a[N]에 대하여 downheap을 수행하지 않는 이유는 이 노드들이 이파리 이므로 
    # 각각의 노드가 힙 크기가 1인 독립적인 최소힙이기 때문이다.
    # 따라서 맨아래 그림같이 최하단의 바로 위층의 마지막 노드인 a[N/2]으로부터 자식 승자와 비교하여 
    # 힙속성에 위배되면 부모와 자식 승자를 교환하는 downheap을 수행한다.
    # 마지막으로 a[1]에 대해 downheap을 수행하면 힙이 완성된다. 
    # 이것을 상향식 힙 만들기(Bottom-up Heap Construction)라고 한다.
    
    def unheap(self,j): # 힙 올라가며 힙속성 회복
        
        while (j//2)>0:

            if a[j][0] < a[j//2][0]:
                a[j],a[j//2] = a[j//2],a[j]
                j = j//2
            
            elif a[j][0] >= a[j//2][0]:
                break
                
            
            
        
        
        
    
    def print_heap(self): # 힙 출력 
        for item in self.a:
            print(item,'', end ='')
        
````

````python
# 4-3 binary_heap 책보고 다시 해봄. 
class BHeap:
    def __init__(self,a): # 이진힙 생성자 
        self.a = a
        self.N = len(a) -1 # 0번 제외   

    def create_heap(self) : # 초기 힙 만들기 heapq.heapify()와 동일 
        for i in range(self.N//2,0, -1):
            self.downheap(i)
            
    def insert(self, key_value): # 삽입연산
        self.N +=1
        self.a.append(key_value) # 새 항목을 힙 마지막에 추가 
        self.upheap(self.N) # 힙 속성을 회복시키기 위해.
        
    def delete_min(self): # 최솟값 삭제  
        if self.N == 0:
            print("힙이 비어있음")
            return None
        minimum = self.a[1]
        self.a[1], self.a[-1] = self.a[-1], self.a[1] # a[1]과 a[N] 교환
        del self.a[-1]
        self.N-=1
        self.downheap(1)
        return minimum
    
    def downheap(self,i): # 힙 내려가며 힙속성 회복
        while 2*i <= self.N:
            #왼쪽 오른쪽 자식 중에서 승자 결정 
            k = 2*i       # 근데 이경우 k< N인뗴 k+1 <N인경우 어떻게 하나?에러아닌가?
            if k < self.N and self.a[k][0] > self.a[k+1][0]:  # 위같은 경우 앞에서 and 하기전에 걸러져서 뒤에 대입도 안하고 끝나는듯.
                k +=1
            if self.a[i][0] < self.a[k][0]: # 힙속성 만족하면 루프 만나기
                break
            self.a[i], self.a[k] = self.a[k], self.a[i]
            i = k
    
    def upheap(self,j): #힙을 올라가며 힙속성 회복
        while j > 1 and self.a[j//2][0] > self.a[j][0]:
            self.a[j],self.a[j//2] = self.a[j//2], self.a[j]
            j = j//2
            
    def print_heap(self): # 힙 출력
        for i in range(1, self.N+1):
            print('[%2d]' % self.a[i][0], self.a[i][1],']', end = "")
        print('\n힙의 크기 = ', self.N)
````

**수행시간**

Insert 연산을 위한 upheap은 삽입된 노드로부터 최대 루트까지 올라가며 부모와 자식 노드를 교환한다.  
delete_min 연산에서는 힙의 마지막 노드를 루트로 이동한 후 downheap을 최하위층의 노드까지 교환해야 하는 경우가 발생한다.  
따라서 힙에서 각 연산의 수행시간은 힙의 높이에 비례한다.  
그런데 힙은 완전이진트리이므로 힙에 N개의 노드가 있으면 그 높이는[log(N+1)]이다.  
결론적으로 각 힙 연산의 수행시간은 O(logN)이다.  

**상향식 힙 만들기**(Bottom-up Heap Construction)의 수행시간 분석을 위해 노드 수가 N인 힙의 각 층에 있는 노드 수를 살펴보자.(간단계산 위해 N = 2^k-1로 가정)  
최하위층 즉 h = 0 인 경우
[N/2]개의 이파리들 있고, 바로 위층(h=1)에는 [N/4]개의 노드가 있다. h=2 에는 [N/8]개의 노드  
즉 h 층에는 N/2^(h+1)개의 노드 가 있다.  
그런데 힙만들기는 h=1인 경우 부터 시작하여 최상위 층의 루트까지 각 노드에 대해 downheap을 수행하므로  
**힙만들기의 수행시간**은 T(N)은 다음 계산을 통해 O(N)임을 알 수 있다.  (N은 전체 노드 수)

레벨순회는 큐 자료구조를 사용해서 구현된다. 
5. 이진트리의 높이는 후위순회 방식으로 계산한다. 
6. 스택없이 이진트리를 순회하기 위해 노드의 None 레퍼런스 대신 다음에 방문할 노드의 레퍼런스를 저장한 이진트리를 스레드 이진트리라고 한다.  
7. 이진트리의 높이 계산, 트리의 각 순회 방식은 트리의 모든 노드들을 방문해야 하므로 각각 O(N)시간이 소요된다.  
8. 우선순위큐는 가장 높은 우선순위를 가진 항목을 접근,삭제,삽입 연산을 지원하는 자료구조이다.
